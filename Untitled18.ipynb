```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
🧬 PHISHING DETECTION SYSTEM v10.0 - IMPROVED & FIXED EDITION
==================================================================
✓ Fixed all bugs and errors from previous version
✓ Improved error handling and fallback mechanisms
✓ Enhanced crawling with better timeout handling
✓ Optimized vector database operations
✓ Better campaign detection algorithm
✓ Improved UI/UX with clearer messages
✓ Added validation and safety checks

INSTALLATION:
pip install playwright sentence-transformers chromadb scikit-learn pandas numpy \
    matplotlib seaborn psutil requests tqdm pillow opencv-python networkx joblib

playwright install chromium
"""

import os
import re
import json
import logging
import hashlib
import warnings
import webbrowser
import requests
import threading
import time
import psutil
import asyncio
from datetime import datetime
from urllib.parse import urlparse
from collections import Counter, deque
from dataclasses import dataclass, asdict, field
from typing import Dict, List, Optional, Tuple, Any
from pathlib import Path

import numpy as np
import pandas as pd
from tqdm import tqdm

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, IsolationForest
from sklearn.metrics import accuracy_score, classification_report
from sklearn.cluster import DBSCAN
from sklearn.manifold import TSNE

# Optional Dependencies
try:
    import cv2
    HAS_CV = True
except: 
    HAS_CV = False

try:
    import networkx as nx
    HAS_NETWORKX = True
except: 
    HAS_NETWORKX = False

try:
    from playwright.async_api import async_playwright
    HAS_PLAYWRIGHT = True
except: 
    HAS_PLAYWRIGHT = False

try:
    from sentence_transformers import SentenceTransformer
    HAS_TRANSFORMERS = True
except: 
    HAS_TRANSFORMERS = False

try:
    import chromadb
    HAS_VECTORDB = True
except: 
    HAS_VECTORDB = False

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")


# ==================== COLORS & UI ====================

class Colors:
    """Terminal colors for beautiful output"""
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    
    @staticmethod
    def print_header(text): 
        print(f"\n{Colors.BOLD}{Colors.HEADER}{'='*80}\n  {text}\n{'='*80}{Colors.ENDC}\n")
    
    @staticmethod
    def print_success(text): 
        print(f"{Colors.OKGREEN}✓ {text}{Colors.ENDC}")
    
    @staticmethod
    def print_error(text): 
        print(f"{Colors.FAIL}✗ {text}{Colors.ENDC}")
    
    @staticmethod
    def print_info(text): 
        print(f"{Colors.OKCYAN}ℹ {text}{Colors.ENDC}")
    
    @staticmethod
    def print_warning(text): 
        print(f"{Colors.WARNING}⚠ {text}{Colors.ENDC}")


# ==================== CONFIGURATION ====================

class Config:
    """System configuration with validation"""
    # Base directories (auto-create on Desktop)
    OUT_DIR = os.path.join(os.path.expanduser("~"), "Desktop", "phishing_detection_v10")
    
    # Try to find dataset in common locations
    POSSIBLE_DATA_PATHS = [
        os.path.join(os.path.expanduser("~"), "Desktop", "datasets", "final_cleaned_preprocessed.csv"),
        os.path.join(os.path.expanduser("~"), "Desktop", "phishing_dataset.csv"),
        os.path.join(os.path.expanduser("~"), "Downloads", "phishing_dataset.csv"),
    ]
    DATA_FILE = None  # Will be set in validate()
    
    MODEL_FILE = os.path.join(OUT_DIR, "phishing_model.pkl")
    
    # Sub-directories
    PLOTS_DIR = os.path.join(OUT_DIR, "analysis_plots")
    REPORTS_DIR = os.path.join(OUT_DIR, "reports")
    THREAT_INTEL_CACHE = os.path.join(OUT_DIR, "threat_intelligence")
    SCREENSHOTS_DIR = os.path.join(OUT_DIR, "screenshots")
    VISUALIZATION_DIR = os.path.join(OUT_DIR, "visualizations")
    CYBER_MAP_DIR = os.path.join(OUT_DIR, "cyber_maps")
    DASHBOARD_DIR = os.path.join(OUT_DIR, "dashboards")
    THREAT_DNA_DIR = os.path.join(OUT_DIR, "threat_dna")
    CRAWL_ARTIFACTS = os.path.join(THREAT_DNA_DIR, "artifacts")
    HAR_LOGS = os.path.join(THREAT_DNA_DIR, "har_logs")
    HTML_DUMPS = os.path.join(THREAT_DNA_DIR, "html_dumps")
    VECTOR_DB_DIR = os.path.join(THREAT_DNA_DIR, "vector_db")
    CAMPAIGN_DATA = os.path.join(THREAT_DNA_DIR, "campaigns")
    
    LOG_FILE = os.path.join(OUT_DIR, "phishing.log")
    
    # Model parameters
    TEST_SIZE = 0.25
    RANDOM_STATE = 42
    HIGH_RISK = 70
    MEDIUM_RISK = 45
    
    # Ensemble weights (optimized)
    WEIGHTS = {
        'rf': 0.20, 
        'gb': 0.20, 
        'adversarial_rf': 0.15, 
        'threat_intel': 0.15, 
        'evasion_detector': 0.10, 
        'threat_dna': 0.20
    }
    
    # Adversarial defense
    ADVERSARIAL_ENABLED = True
    ADVERSARIAL_NOISE_EPSILON = 0.1
    ADVERSARIAL_TRAINING_RATIO = 0.3
    EVASION_THRESHOLD = 0.7
    
    # Monitoring
    MONITORING_INTERVAL = 5
    MAX_HISTORY = 100
    CHAT_HISTORY_SIZE = 50
    
    # Threat DNA (optimized timeouts)
    CRAWL_TIMEOUT = 45000  # Increased to 45 seconds
    CRAWL_WAIT_TIME = 3000  # Reduced to 3 seconds
    ENABLE_SCREENSHOTS = True
    ENABLE_HAR_CAPTURE = False  # Disabled by default (causes issues)
    ENABLE_DOM_SNAPSHOT = True
    TEXT_EMBEDDING_MODEL = "all-MiniLM-L6-v2"
    EMBEDDING_DIM = 384
    SIMILARITY_THRESHOLD = 0.85
    CAMPAIGN_CLUSTERING_EPS = 0.3
    CAMPAIGN_MIN_SAMPLES = 3
    VIZ_DPI = 150
    
    @classmethod
    def validate(cls):
        """Validate and setup configuration"""
        # Create all directories
        for directory in [cls.OUT_DIR, cls.PLOTS_DIR, cls.REPORTS_DIR, cls.THREAT_INTEL_CACHE,
                          cls.SCREENSHOTS_DIR, cls.VISUALIZATION_DIR, cls.CYBER_MAP_DIR,
                          cls.DASHBOARD_DIR, cls.THREAT_DNA_DIR, cls.CRAWL_ARTIFACTS,
                          cls.HAR_LOGS, cls.HTML_DUMPS, cls.VECTOR_DB_DIR, cls.CAMPAIGN_DATA]:
            os.makedirs(directory, exist_ok=True)
        
        # Find dataset file
        for path in cls.POSSIBLE_DATA_PATHS:
            if os.path.exists(path):
                cls.DATA_FILE = path
                break
        
        return cls.DATA_FILE is not None


# Validate configuration
Config.validate()

# Setup logging
logging.basicConfig(
    level=logging.INFO, 
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler(Config.LOG_FILE), 
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)


# ==================== THREAT DNA ====================

@dataclass
class ThreatDNA:
    """🧬 Complete threat signature"""
    url: str
    timestamp: datetime
    threat_id: str = field(default="")
    screenshot_path: Optional[str] = None
    html_content: Optional[str] = None
    console_logs: List[str] = field(default_factory=list)
    dom_snapshot: Optional[Dict] = None
    network_requests: List[Dict] = field(default_factory=list)
    text_content: Optional[str] = None
    extracted_links: List[str] = field(default_factory=list)
    extracted_forms: List[Dict] = field(default_factory=list)
    javascript_behavior: Dict = field(default_factory=dict)
    text_embedding: Optional[np.ndarray] = None
    image_embedding: Optional[np.ndarray] = None
    dom_embedding: Optional[np.ndarray] = None
    combined_embedding: Optional[np.ndarray] = None
    page_title: Optional[str] = None
    response_time_ms: Optional[float] = None
    risk_score: float = 0.0
    threat_category: Optional[str] = None
    campaign_id: Optional[str] = None
    similar_threats: List[str] = field(default_factory=list)
    
    def __post_init__(self):
        if not self.threat_id:
            self.threat_id = self.generate_threat_id()
    
    def generate_threat_id(self) -> str:
        """Generate unique threat identifier"""
        components = f"{self.url}{self.timestamp.isoformat()}".encode('utf-8')
        return f"THR-{hashlib.sha256(components).hexdigest()[:16].upper()}"
    
    def get_fingerprint(self) -> str:
        """Generate unique fingerprint"""
        components = [
            self.url,
            self.text_content or "",
            str(self.dom_snapshot) if self.dom_snapshot else "",
            str(self.javascript_behavior)
        ]
        return hashlib.sha256("".join(components).encode('utf-8')).hexdigest()
    
    def to_dict(self) -> Dict:
        """Convert to dictionary for JSON serialization"""
        data = asdict(self)
        # Convert numpy arrays to lists
        for key in ['text_embedding', 'image_embedding', 'dom_embedding', 'combined_embedding']:
            if data[key] is not None and isinstance(data[key], np.ndarray):
                data[key] = data[key].tolist()
        data['timestamp'] = self.timestamp.isoformat()
        return data


# ==================== THREAT DNA CRAWLER ====================

class ThreatDNACrawler:
    """🕷️ Automated web crawler with improved error handling"""
    
    def __init__(self):
        self.storage_dir = Path(Config.CRAWL_ARTIFACTS)
        self.screenshots_dir = Path(Config.SCREENSHOTS_DIR)
        self.html_dir = Path(Config.HTML_DUMPS)
        logger.info("✓ Threat DNA Crawler initialized")
    
    async def crawl_url(self, url: str) -> Optional[ThreatDNA]:
        """Crawl URL in isolated browser with comprehensive error handling"""
        if not HAS_PLAYWRIGHT:
            Colors.print_error("Playwright not installed!")
            return None
        
        Colors.print_info(f"🕷️ Crawling: {url[:60]}...")
        start_time = datetime.now()
        dna = ThreatDNA(url=url, timestamp=start_time)
        
        browser = None
        try:
            async with async_playwright() as p:
                # Launch browser with optimized settings
                browser = await p.chromium.launch(
                    headless=True,
                    args=[
                        '--no-sandbox',
                        '--disable-setuid-sandbox',
                        '--disable-dev-shm-usage',
                        '--disable-gpu',
                        '--disable-software-rasterizer',
                        '--disable-dev-shm-usage'
                    ]
                )
                
                # Create context with timeout
                context = await browser.new_context(
                    viewport={'width': 1920, 'height': 1080},
                    user_agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
                    ignore_https_errors=True,
                    java_script_enabled=True
                )
                
                # Set default timeout
                context.set_default_timeout(Config.CRAWL_TIMEOUT)
                
                # Create page
                page = await context.new_page()
                
                # Setup event listeners
                page.on("console", lambda msg: dna.console_logs.append(
                    f"[{msg.type}] {msg.text}"
                ))
                page.on("request", lambda req: dna.network_requests.append({
                    'url': req.url,
                    'method': req.method,
                    'timestamp': datetime.now().isoformat()
                }))
                
                # Navigate to URL with error handling
                try:
                    response = await page.goto(
                        url,
                        timeout=Config.CRAWL_TIMEOUT,
                        wait_until='domcontentloaded'  # Changed from networkidle
                    )
                    
                    if response is None:
                        raise Exception("No response received")
                    
                    dna.response_time_ms = (datetime.now() - start_time).total_seconds() * 1000
                    
                    # Wait for page to settle (shorter timeout)
                    try:
                        await page.wait_for_load_state('networkidle', timeout=Config.CRAWL_WAIT_TIME)
                    except:
                        pass  # Continue even if network doesn't settle
                    
                    # Extract page data
                    dna.page_title = await page.title()
                    dna.html_content = await page.content()
                    
                    # Take screenshot
                    if Config.ENABLE_SCREENSHOTS:
                        try:
                            screenshot_path = self.screenshots_dir / f"{dna.threat_id}.png"
                            await page.screenshot(
                                path=str(screenshot_path), 
                                full_page=False,  # Only visible area
                                timeout=10000
                            )
                            dna.screenshot_path = str(screenshot_path)
                        except Exception as e:
                            logger.warning(f"Screenshot failed: {e}")
                    
                    # Extract DOM snapshot
                    if Config.ENABLE_DOM_SNAPSHOT:
                        try:
                            dna.dom_snapshot = await page.evaluate('''() => {
                                try {
                                    return {
                                        title: document.title || '',
                                        forms: Array.from(document.forms || []).slice(0, 10).map(f => ({
                                            action: f.action || '',
                                            method: f.method || '',
                                            inputs: Array.from(f.elements || []).slice(0, 20).map(e => ({
                                                type: e.type || '',
                                                name: e.name || ''
                                            }))
                                        })),
                                        links: Array.from(document.links || []).slice(0, 50).map(l => l.href || ''),
                                        scripts: Array.from(document.scripts || []).slice(0, 20).map(s => s.src || ''),
                                        iframes: Array.from(document.querySelectorAll('iframe') || []).slice(0, 10).map(i => i.src || '')
                                    }
                                } catch(e) {
                                    return {title: '', forms: [], links: [], scripts: [], iframes: []};
                                }
                            }''')
                        except Exception as e:
                            logger.warning(f"DOM snapshot failed: {e}")
                            dna.dom_snapshot = {'forms': [], 'links': [], 'scripts': [], 'iframes': []}
                    
                    # Extract text content
                    try:
                        text_content = await page.evaluate('() => document.body ? document.body.innerText : ""')
                        dna.text_content = text_content[:10000] if text_content else ""
                    except:
                        dna.text_content = ""
                    
                    # Extract links and forms
                    if dna.dom_snapshot:
                        dna.extracted_links = dna.dom_snapshot.get('links', [])
                        dna.extracted_forms = dna.dom_snapshot.get('forms', [])
                    
                    # JavaScript behavior analysis
                    try:
                        dna.javascript_behavior = await page.evaluate('''() => {
                            try {
                                return {
                                    hasOnLoad: !!window.onload,
                                    hasLocalStorage: typeof(Storage) !== 'undefined',
                                    hasGeolocation: 'geolocation' in navigator,
                                    windowProperties: Object.keys(window).length,
                                    hasAlert: typeof alert !== 'undefined',
                                    hasPrompt: typeof prompt !== 'undefined'
                                }
                            } catch(e) {
                                return {};
                            }
                        }''')
                    except:
                        dna.javascript_behavior = {}
                    
                    Colors.print_success("✓ Crawl completed successfully")
                
                except Exception as e:
                    error_str = str(e)
                    logger.error(f"Crawl error: {e}")
                    
                    # Provide specific error messages
                    if "ERR_NAME_NOT_RESOLVED" in error_str or "net::ERR_NAME_NOT_RESOLVED" in error_str:
                        Colors.print_error("DNS resolution failed - domain doesn't exist")
                    elif "ERR_CONNECTION_REFUSED" in error_str:
                        Colors.print_error("Connection refused - server is down")
                    elif "ERR_CONNECTION_TIMED_OUT" in error_str or "Timeout" in error_str:
                        Colors.print_error("Connection timeout - server too slow")
                    elif "ERR_SSL_PROTOCOL_ERROR" in error_str or "SSL" in error_str:
                        Colors.print_error("SSL/TLS error - invalid certificate")
                    else:
                        Colors.print_error(f"Navigation failed: {str(e)[:80]}")
                    
                    return None
                
                # Save HTML
                if dna.html_content:
                    try:
                        html_path = self.html_dir / f"{dna.threat_id}.html"
                        with open(html_path, 'w', encoding='utf-8', errors='ignore') as f:
                            f.write(dna.html_content)
                    except Exception as e:
                        logger.warning(f"HTML save failed: {e}")
                
                await context.close()
                await browser.close()
            
            return dna
        
        except Exception as e:
            logger.error(f"Fatal crawl error: {e}")
            if browser:
                try:
                    await browser.close()
                except:
                    pass
            return None
    
    def crawl_url_sync(self, url: str) -> Optional[ThreatDNA]:
        """Synchronous wrapper with proper event loop handling"""
        try:
            # Try to create new event loop
            try:
                # Close any existing loop first
                try:
                    old_loop = asyncio.get_event_loop()
                    if old_loop.is_running():
                        # Can't use running loop from main thread
                        Colors.print_warning("Event loop already running - creating new thread")
                        import concurrent.futures
                        with concurrent.futures.ThreadPoolExecutor() as executor:
                            future = executor.submit(self._run_in_new_loop, url)
                            return future.result(timeout=60)
                except RuntimeError:
                    pass
                
                # Create fresh event loop
                loop = asyncio.new_event_loop()
                asyncio.set_event_loop(loop)
                try:
                    result = loop.run_until_complete(self.crawl_url(url))
                    return result
                finally:
                    try:
                        pending = asyncio.all_tasks(loop)
                        for task in pending:
                            task.cancel()
                        loop.run_until_complete(asyncio.gather(*pending, return_exceptions=True))
                    except:
                        pass
                    try:
                        loop.close()
                    except:
                        pass
            except Exception as e:
                logger.error(f"Event loop error: {e}")
                return None
        except Exception as e:
            logger.error(f"Sync crawl error: {e}")
            Colors.print_error(f"Crawl error: {str(e)[:100]}")
            return None
    
    def _run_in_new_loop(self, url: str) -> Optional[ThreatDNA]:
        """Run crawl in a completely new event loop in new thread"""
        loop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)
        try:
            return loop.run_until_complete(self.crawl_url(url))
        finally:
            loop.close()


# ==================== EMBEDDING GENERATOR ====================

class EmbeddingGenerator:
    """🔢 Multi-modal embedding generation with error handling"""
    
    def __init__(self):
        self.text_model = None
        if HAS_TRANSFORMERS:
            try:
                Colors.print_info("Loading text embedding model...")
                self.text_model = SentenceTransformer(Config.TEXT_EMBEDDING_MODEL)
                Colors.print_success("✓ Text embedding model loaded")
            except Exception as e:
                Colors.print_warning(f"Text model load failed: {e}")
    
    def generate_text_embedding(self, text: str) -> Optional[np.ndarray]:
        """Generate text embedding with validation"""
        if not self.text_model or not text or len(text.strip()) < 10:
            return None
        try:
            # Limit text length
            text = text[:5000]
            embedding = self.text_model.encode(text, show_progress_bar=False, convert_to_numpy=True)
            return np.array(embedding)
        except Exception as e:
            logger.warning(f"Text embedding failed: {e}")
            return None
    
    def generate_image_embedding(self, image_path: str) -> Optional[np.ndarray]:
        """Generate image embedding (color histogram)"""
        if not HAS_CV or not image_path or not os.path.exists(image_path):
            return None
        try:
            img = cv2.imread(image_path)
            if img is None or img.size == 0:
                return None
            
            # Resize if too large
            max_size = 1024
            h, w = img.shape[:2]
            if max(h, w) > max_size:
                scale = max_size / max(h, w)
                img = cv2.resize(img, (int(w * scale), int(h * scale)))
            
            # Calculate color histograms
            hist_r = cv2.calcHist([img], [0], None, [256], [0, 256]).flatten()
            hist_g = cv2.calcHist([img], [1], None, [256], [0, 256]).flatten()
            hist_b = cv2.calcHist([img], [2], None, [256], [0, 256]).flatten()
            
            # Concatenate and normalize
            hist = np.concatenate([hist_r, hist_g, hist_b])
            hist = hist / (hist.sum() + 1e-7)
            
            # Reduce to embedding dimension
            if len(hist) > Config.EMBEDDING_DIM:
                step = len(hist) // Config.EMBEDDING_DIM
                hist = hist[::step][:Config.EMBEDDING_DIM]
            if len(hist) < Config.EMBEDDING_DIM:
                hist = np.pad(hist, (0, Config.EMBEDDING_DIM - len(hist)))
            
            return hist
        except Exception as e:
            logger.warning(f"Image embedding failed: {e}")
            return None
    
    def generate_dom_embedding(self, dom_snapshot: Dict) -> Optional[np.ndarray]:
        """Generate DOM structure embedding"""
        if not dom_snapshot:
            return None
        try:
            features = [
                len(dom_snapshot.get('forms', [])),
                len(dom_snapshot.get('links', [])),
                len(dom_snapshot.get('scripts', [])),
                len(dom_snapshot.get('iframes', [])),
                sum(1 for f in dom_snapshot.get('forms', []) 
                    if f.get('method', '').lower() == 'post'),
                sum(1 for l in dom_snapshot.get('links', []) 
                    if str(l).startswith('javascript:')),
                sum(1 for s in dom_snapshot.get('scripts', []) if s),  # Non-empty scripts
                sum(1 for i in dom_snapshot.get('iframes', []) if i)   # Non-empty iframes
            ]
            
            # Pad to embedding dimension
            embedding = np.array(features + [0] * (Config.EMBEDDING_DIM - len(features)))
            return embedding[:Config.EMBEDDING_DIM]
        except Exception as e:
            logger.warning(f"DOM embedding failed: {e}")
            return None
    
    def generate_combined_embedding(self, dna: ThreatDNA) -> Optional[np.ndarray]:
        """Combine all embeddings with weighted average"""
        embeddings = []
        weights = []
        
        if dna.text_embedding is not None:
            embeddings.append(dna.text_embedding)
            weights.append(0.5)  # Text is most important
        
        if dna.image_embedding is not None:
            embeddings.append(dna.image_embedding)
            weights.append(0.3)  # Image second
        
        if dna.dom_embedding is not None:
            embeddings.append(dna.dom_embedding)
            weights.append(0.2)  # DOM third
        
        if not embeddings:
            return None
        
        try:
            # Weighted average
            weights = np.array(weights)
            weights = weights / weights.sum()
            combined = np.average(embeddings, axis=0, weights=weights)
            
            # Normalize
            norm = np.linalg.norm(combined)
            return combined / norm if norm > 0 else combined
        except Exception as e:
            logger.warning(f"Combined embedding failed: {e}")
            return None
    
    def enrich_threat_dna(self, dna: ThreatDNA) -> ThreatDNA:
        """Enrich ThreatDNA with all embeddings"""
        try:
            if dna.text_content and len(dna.text_content.strip()) > 10:
                dna.text_embedding = self.generate_text_embedding(dna.text_content)
            
            if dna.screenshot_path and os.path.exists(dna.screenshot_path):
                dna.image_embedding = self.generate_image_embedding(dna.screenshot_path)
            
            if dna.dom_snapshot:
                dna.dom_embedding = self.generate_dom_embedding(dna.dom_snapshot)
            
            dna.combined_embedding = self.generate_combined_embedding(dna)
        except Exception as e:
            logger.warning(f"DNA enrichment error: {e}")
        
        return dna


# ==================== VECTOR DATABASE ====================

class ThreatVectorDB:
    """🗄️ Vector database with improved error handling"""
    
    def __init__(self):
        self.client = None
        self.collection = None
        
        if HAS_VECTORDB:
            try:
                self.client = chromadb.PersistentClient(path=Config.VECTOR_DB_DIR)
                self.collection = self.client.get_or_create_collection(
                    name="threat_dna",
                    metadata={"description": "Phishing threat DNA vectors"}
                )
                Colors.print_success(f"✓ Vector DB initialized ({self.collection.count()} entries)")
            except Exception as e:
                Colors.print_warning(f"Vector DB init failed: {e}")
                self.client = None
                self.collection = None
    
    def add_threat(self, dna: ThreatDNA) -> bool:
        """Add threat to vector database with validation"""
        if not self.collection or dna.combined_embedding is None:
            return False
        
        try:
            # Check if already exists
            try:
                existing = self.collection.get(ids=[dna.threat_id])
                if existing['ids']:
                    logger.info(f"Threat {dna.threat_id} already in DB, updating...")
                    self.collection.delete(ids=[dna.threat_id])
            except:
                pass
            
            # Add new entry
            self.collection.add(
                ids=[dna.threat_id],
                embeddings=[dna.combined_embedding.tolist()],
                metadatas=[{
                    'url': dna.url[:500],  # Limit length
                    'timestamp': dna.timestamp.isoformat(),
                    'risk_score': float(dna.risk_score),
                    'threat_category': dna.threat_category or 'unknown'
                }]
            )
            return True
        except Exception as e:
            logger.error(f"Vector DB add failed: {e}")
            return False
    
    def find_similar_threats(self, dna: ThreatDNA, top_k: int = 5) -> List[Dict]:
        """Find similar threats with validation"""
        if not self.collection or dna.combined_embedding is None:
            return []
        
        try:
            # Query with increased results to filter self
            results = self.collection.query(
                query_embeddings=[dna.combined_embedding.tolist()],
                n_results=min(top_k + 1, self.collection.count())
            )
            
            similar_threats = []
            if results['ids'] and len(results['ids']) > 0:
                for i, threat_id in enumerate(results['ids'][0]):
                    if threat_id != dna.threat_id:  # Exclude self
                        similarity = 1 - results['distances'][0][i]
                        if similarity > 0.5:  # Only meaningful similarities
                            similar_threats.append({
                                'threat_id': threat_id,
                                'similarity': similarity,
                                'metadata': results['metadatas'][0][i]
                            })
            
            return similar_threats[:top_k]
        except Exception as e:
            logger.error(f"Vector search failed: {e}")
            return []
    
    def get_threat_count(self) -> int:
        """Get total number of threats"""
        if not self.collection:
            return 0
        try:
            return self.collection.count()
        except:
            return 0


# ==================== CAMPAIGN DETECTOR ====================

class CampaignDetector:
    """🎯 Detect phishing campaigns with improved clustering"""
    
    def __init__(self):
        self.campaigns = {}
        self.next_campaign_id = 1
    
    def detect_campaigns(self, threat_dnas: List[ThreatDNA]) -> Dict[str, List[ThreatDNA]]:
        """Cluster threats into campaigns using DBSCAN with validation"""
        if len(threat_dnas) < Config.CAMPAIGN_MIN_SAMPLES:
            return {'UNCLUSTERED': threat_dnas}
        
        # Extract embeddings
        embeddings = []
        valid_dnas = []
        for dna in threat_dnas:
            if dna.combined_embedding is not None and len(dna.combined_embedding) == Config.EMBEDDING_DIM:
                embeddings.append(dna.combined_embedding)
                valid_dnas.append(dna)
        
        if len(embeddings) < Config.CAMPAIGN_MIN_SAMPLES:
            return {'UNCLUSTERED': threat_dnas}
        
        try:
            # DBSCAN clustering with adaptive parameters
            X = np.array(embeddings)
            
            # Normalize embeddings
            norms = np.linalg.norm(X, axis=1, keepdims=True)
            X = X / (norms + 1e-8)
            
            # Adaptive EPS based on data size
            eps = Config.CAMPAIGN_CLUSTERING_EPS
            if len(embeddings) < 10:
                eps = 0.4  # More lenient for small datasets
            
            clustering = DBSCAN(
                eps=eps,
                min_samples=min(Config.CAMPAIGN_MIN_SAMPLES, len(embeddings) // 3),
                metric='cosine'
            ).fit(X)
            
            # Group by campaign
            campaigns = {}
            for idx, label in enumerate(clustering.labels_):
                if label == -1:
                    campaign_id = 'UNCLUSTERED'
                else:
                    campaign_id = f'CAMPAIGN-{label + 1}'
                
                if campaign_id not in campaigns:
                    campaigns[campaign_id] = []
                
                campaigns[campaign_id].append(valid_dnas[idx])
                valid_dnas[idx].campaign_id = campaign_id
            
            # Add unclustered DNAs that had no embeddings
            unclustered_dnas = [d for d in threat_dnas if d not in valid_dnas]
            if unclustered_dnas:
                if 'UNCLUSTERED' not in campaigns:
                    campaigns['UNCLUSTERED'] = []
                campaigns['UNCLUSTERED'].extend(unclustered_dnas)
            
            return campaigns
        
        except Exception as e:
            logger.error(f"Campaign detection failed: {e}")
            return {'UNCLUSTERED': threat_dnas}
    
    def generate_campaign_report(self, campaigns: Dict[str, List[ThreatDNA]]) -> str:
        """Generate detailed campaign analysis report"""
        report = f"\n{'='*80}\n🎯 CAMPAIGN DETECTION REPORT\n{'='*80}\n\n"
        report += f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n"
        report += f"Total Campaigns Detected: {len([c for c in campaigns if c != 'UNCLUSTERED'])}\n"
        report += f"Unclustered Threats: {len(campaigns.get('UNCLUSTERED', []))}\n\n"
        
        # Sort campaigns by size
        sorted_campaigns = sorted(
            [(k, v) for k, v in campaigns.items() if k != 'UNCLUSTERED'],
            key=lambda x: len(x[1]),
            reverse=True
        )
        
        for campaign_id, threats in sorted_campaigns:
            report += f"\n{campaign_id} ({len(threats)} threats)\n"
            report += f"{'='*60}\n"
            
            if threats:
                avg_risk = np.mean([t.risk_score for t in threats])
                report += f"Average Risk Score: {avg_risk:.1f}/100\n"
                
                timestamps = [t.timestamp for t in threats if t.timestamp]
                if timestamps:
                    report += f"First Seen: {min(timestamps).strftime('%Y-%m-%d %H:%M:%S')}\n"
                    report += f"Last Seen: {max(timestamps).strftime('%Y-%m-%d %H:%M:%S')}\n"
                
                report += f"\nURLs:\n"
                for t in threats[:10]:
                    report += f"  - {t.url[:70]}... (Risk: {t.risk_score}/100)\n"
                if len(threats) > 10:
                    report += f"  ... and {len(threats) - 10} more\n"
        
        if 'UNCLUSTERED' in campaigns and campaigns['UNCLUSTERED']:
            report += f"\n\nUNCLUSTERED THREATS ({len(campaigns['UNCLUSTERED'])})\n"
            report += f"{'='*60}\n"
            for t in campaigns['UNCLUSTERED'][:5]:
                report += f"  - {t.url[:70]}... (Risk: {t.risk_score}/100)\n"
            if len(campaigns['UNCLUSTERED']) > 5:
                report += f"  ... and {len(campaigns['UNCLUSTERED']) - 5} more\n"
        
        return report


# ==================== SUPPORTING CLASSES ====================

class SecurityChatAssistant:
    """🤖 AI Security Copilot"""
    
    def __init__(self):
        self.chat_history = deque(maxlen=Config.CHAT_HISTORY_SIZE)
        self.context = {'recent_detections': [], 'threat_stats': {}, 'system_health': {}}
    
    def process_query(self, query: str) -> str:
        """Process user query and generate response"""
        query_lower = query.lower().strip()
        
        self.chat_history.append({
            'timestamp': datetime.now(),
            'query': query,
            'type': 'user'
        })
        
        response = self._generate_response(query_lower)
        
        self.chat_history.append({
            'timestamp': datetime.now(),
            'response': response,
            'type': 'assistant'
        })
        
        return response
    
    def _generate_response(self, query: str) -> str:
        """Generate AI response based on query"""
        if 'statistics' in query or 'stats' in query:
            total = len(self.context.get('recent_detections', []))
            phishing = sum(1 for d in self.context.get('recent_detections', []) 
                          if d.get('is_phishing'))
            return f"📊 Statistics: {total} total scans, {phishing} phishing detected ({(phishing/total*100) if total > 0 else 0:.1f}%)"
        
        elif 'help' in query:
            return "🤖 I can help with: statistics, threat analysis, protection tips, system status, or general cybersecurity questions."
        
        elif 'threat' in query or 'phishing' in query:
            return "🎣 Phishing is a cyber attack that uses disguised communication to steal sensitive information. Always verify sources before clicking links!"
        
        elif 'status' in query or 'health' in query:
            health = self.context.get('system_health', {})
            return f"⚙️ System Status: Models ✓ | CPU: {health.get('cpu', 0):.1f}% | Memory: {health.get('memory', 0):.1f}%"
        
        elif 'protect' in query or 'safe' in query:
            return "🛡️ Protection tips: 1) Check sender address 2) Hover over links 3) Look for HTTPS 4) Don't rush 5) Enable 2FA"
        
        elif 'latest' in query or 'recent' in query:
            recent = self.context.get('recent_detections', [])[-5:]
            if recent:
                return f"📋 Last {len(recent)} scans: {sum(1 for d in recent if d.get('is_phishing'))} threats detected"
            return "No recent scans available"
        
        else:
            return "I'm your AI security assistant! Ask me about statistics, threats, protection tips, or system status."
    
    def update_context(self, detections=None, threat_stats=None, system_health=None):
        """Update assistant context"""
        if detections: self.context['recent_detections'] = detections
        if threat_stats: self.context['threat_stats'] = threat_stats
        if system_health: self.context['system_health'] = system_health


class RealTimeMonitor:
    """Real-time monitoring system"""
    
    def __init__(self):
        self.detection_history = deque(maxlen=Config.MAX_HISTORY)
        self.threat_timeline = deque(maxlen=Config.MAX_HISTORY)
        self.system_metrics = deque(maxlen=Config.MAX_HISTORY)
        self.is_monitoring = False
        self.monitor_thread = None
    
    def start_monitoring(self):
        """Start background monitoring"""
        if self.is_monitoring:
            return
        
        self.is_monitoring = True
        self.monitor_thread = threading.Thread(target=self._monitoring_loop, daemon=True)
        self.monitor_thread.start()
        Colors.print_success("✓ Real-time monitoring started")
    
    def stop_monitoring(self):
        """Stop background monitoring"""
        self.is_monitoring = False
        if self.monitor_thread:
            self.monitor_thread.join(timeout=2)
    
    def _monitoring_loop(self):
        """Background monitoring loop"""
        while self.is_monitoring:
            try:
                metrics = {
                    'timestamp': datetime.now(),
                    'cpu_percent': psutil.cpu_percent(interval=1),
                    'memory_percent': psutil.virtual_memory().percent,
                    'active_threats': len([d for d in self.detection_history if d.get('is_phishing')])
                }
                self.system_metrics.append(metrics)
                time.sleep(Config.MONITORING_INTERVAL)
            except Exception as e:
                logger.error(f"Monitoring error: {e}")
                time.sleep(Config.MONITORING_INTERVAL)
    
    def log_detection(self, detection_result: Dict):
        """Log a detection event"""
        event = {
            'timestamp': datetime.now(),
            'url': detection_result.get('url', '')[:50],
            'is_phishing': detection_result.get('is_phishing', False),
            'risk_score': detection_result.get('risk_score', 0),
            'risk_level': detection_result.get('risk_level', 'UNKNOWN')
        }
        self.detection_history.append(event)
        if event['is_phishing']:
            self.threat_timeline.append(event)
    
    def get_current_stats(self) -> Dict:
        """Get current monitoring statistics"""
        total = len(self.detection_history)
        phishing = sum(1 for d in self.detection_history if d['is_phishing'])
        latest_metrics = self.system_metrics[-1] if self.system_metrics else {}
        
        return {
            'total_scans': total,
            'phishing_detected': phishing,
            'safe_urls': total - phishing,
            'detection_rate': (phishing/total*100) if total > 0 else 0,
            'cpu_usage': latest_metrics.get('cpu_percent', 0),
            'memory_usage': latest_metrics.get('memory_percent', 0),
            'active_threats': latest_metrics.get('active_threats', 0)
        }


class CyberMapVisualizer:
    """3D Cyber threat map generator"""
    
    def __init__(self):
        self.country_codes = {
            'US': (37.0902, -95.7129), 'CN': (35.8617, 104.1954), 'RU': (61.5240, 105.3188),
            'BR': (-14.2350, -51.9253), 'IN': (20.5937, 78.9629), 'UK': (55.3781, -3.4360),
            'DE': (51.1657, 10.4515), 'FR': (46.2276, 2.2137), 'CA': (56.1304, -106.3468),
            'AU': (-25.2744, 133.7751), 'JP': (36.2048, 138.2529), 'Unknown': (0, 0)
        }
    
    def create_3d_cyber_map(self, threat_data: List[Dict]) -> Optional[str]:
        """Create 3D visualization of threats"""
        if not threat_data:
            return None
        
        try:
            fig = plt.figure(figsize=(16, 12))
            
            # 3D Globe
            ax1 = fig.add_subplot(221, projection='3d')
            
            # Generate sphere
            u = np.linspace(0, 2 * np.pi, 50)
            v = np.linspace(0, np.pi, 50)
            x = np.outer(np.cos(u), np.sin(v))
            y = np.outer(np.sin(u), np.sin(v))
            z = np.outer(np.ones(np.size(u)), np.cos(v))
            ax1.plot_surface(x, y, z, color='lightblue', alpha=0.3, linewidth=0)
            
            # Plot threats
            for threat in threat_data[-50:]:  # Last 50 threats
                lat, lon = self._get_threat_location(threat)
                lat_rad, lon_rad = np.radians(lat), np.radians(lon)
                threat_x = np.cos(lat_rad) * np.cos(lon_rad)
                threat_y = np.cos(lat_rad) * np.sin(lon_rad)
                threat_z = np.sin(lat_rad)
                
                risk = threat.get('risk_score', 50) / 100
                color = 'red' if risk > 0.7 else 'orange' if risk > 0.45 else 'yellow'
                ax1.scatter([threat_x], [threat_y], [threat_z], 
                           c=color, s=50 + risk * 100, alpha=0.8, edgecolors='black')
            
            ax1.set_title('3D Global Threat Map', fontweight='bold', fontsize=12)
            ax1.set_box_aspect([1,1,1])
            
            # Country Distribution
            ax2 = fig.add_subplot(222)
            if threat_data:
                countries = [self._guess_country(t) for t in threat_data]
                country_counts = Counter(countries)
                
                # Top 10 countries
                top_countries = dict(sorted(country_counts.items(), 
                                          key=lambda x: x[1], reverse=True)[:10])
                
                colors = plt.cm.YlOrRd(np.linspace(0.3, 1, len(top_countries)))
                ax2.barh(list(top_countries.keys()), list(top_countries.values()), color=colors)
                ax2.set_xlabel('Threat Count')
                ax2.set_title('Top 10 Threat Origins', fontweight='bold')
                ax2.grid(axis='x', alpha=0.3)
            
            # Risk Distribution
            ax3 = fig.add_subplot(223)
            risk_scores = [t.get('risk_score', 0) for t in threat_data]
            ax3.hist(risk_scores, bins=20, color='coral', edgecolor='black', alpha=0.7)
            ax3.axvline(x=Config.HIGH_RISK, color='red', linestyle='--', label='High Risk')
            ax3.axvline(x=Config.MEDIUM_RISK, color='orange', linestyle='--', label='Medium Risk')
            ax3.set_xlabel('Risk Score')
            ax3.set_ylabel('Frequency')
            ax3.set_title('Risk Score Distribution', fontweight='bold')
            ax3.legend()
            ax3.grid(alpha=0.3)
            
            # Timeline
            ax4 = fig.add_subplot(224)
            if len(threat_data) > 1:
                timestamps = [t.get('timestamp', datetime.now()) for t in threat_data]
                risks = [t.get('risk_score', 0) for t in threat_data]
                
                colors = ['red' if r > Config.HIGH_RISK else 'orange' if r > Config.MEDIUM_RISK else 'green' 
                         for r in risks]
                ax4.scatter(range(len(timestamps)), risks, c=colors, alpha=0.6, s=50)
                ax4.plot(range(len(timestamps)), risks, 'gray', alpha=0.3, linewidth=1)
                ax4.axhline(y=Config.HIGH_RISK, color='red', linestyle='--', alpha=0.5)
                ax4.axhline(y=Config.MEDIUM_RISK, color='orange', linestyle='--', alpha=0.5)
                ax4.set_xlabel('Detection Sequence')
                ax4.set_ylabel('Risk Score')
                ax4.set_title('Threat Timeline', fontweight='bold')
                ax4.grid(alpha=0.3)
            
            plt.suptitle('🌍 CYBER THREAT MAP - Real-Time Analysis', 
                        fontsize=16, fontweight='bold')
            plt.tight_layout()
            
            filename = f'cyber_map_3d_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png'
            filepath = os.path.join(Config.CYBER_MAP_DIR, filename)
            plt.savefig(filepath, dpi=Config.VIZ_DPI, bbox_inches='tight')
            plt.close()
            
            return filepath
        
        except Exception as e:
            logger.error(f"3D map error: {e}")
            return None
    
    def _get_threat_location(self, threat: Dict) -> Tuple[float, float]:
        """Get threat location coordinates"""
        country = self._guess_country(threat)
        return self.country_codes.get(country, (0, 0))
    
    def _guess_country(self, threat: Dict) -> str:
        """Guess country from URL"""
        url = threat.get('url', '').lower()
        
        # TLD-based detection
        tld_map = {
            '.cn': 'CN', '.ru': 'RU', '.br': 'BR', '.in': 'IN',
            '.uk': 'UK', '.de': 'DE', '.fr': 'FR', '.ca': 'CA',
            '.au': 'AU', '.jp': 'JP'
        }
        
        for tld, country in tld_map.items():
            if tld in url:
                return country
        
        # Default to US for common TLDs
        if any(tld in url for tld in ['.com', '.net', '.org']):
            return 'US'
        
        return 'Unknown'


class AdversarialDefense:
    """Adversarial attack defense system"""
    
    def __init__(self):
        self.attack_history = deque(maxlen=1000)
        self.evasion_attempts_detected = 0
    
    def generate_adversarial_samples(self, X: pd.DataFrame, y: np.ndarray, 
                                    epsilon: float = None) -> pd.DataFrame:
        """Generate adversarial samples for training"""
        if epsilon is None:
            epsilon = Config.ADVERSARIAL_NOISE_EPSILON
        
        X_adv = X.copy()
        for i in range(len(X_adv)):
            noise = np.random.normal(0, epsilon, X_adv.shape[1])
            X_adv.iloc[i] += noise
            # Clip to valid ranges
            X_adv.iloc[i] = X_adv.iloc[i].clip(X.min(), X.max())
        
        return X_adv
    
    def train_adversarial_robust_model(self, X_train: pd.DataFrame, y_train: np.ndarray,
                                      X_test: pd.DataFrame, y_test: np.ndarray):
        """Train model robust to adversarial attacks"""
        try:
            n_adversarial = int(len(X_train) * Config.ADVERSARIAL_TRAINING_RATIO)
            indices = np.random.choice(len(X_train), n_adversarial, replace=False)
            
            X_adv = self.generate_adversarial_samples(
                X_train.iloc[indices], 
                y_train[indices]
            )
            
            X_combined = pd.concat([X_train, X_adv], ignore_index=True)
            y_combined = np.concatenate([y_train, y_train[indices]])
            
            model = RandomForestClassifier(
                n_estimators=300,
                max_depth=25,
                min_samples_split=5,
                min_samples_leaf=2,
                random_state=Config.RANDOM_STATE,
                n_jobs=-1
            )
            model.fit(X_combined, y_combined)
            
            clean_score = model.score(X_test, y_test)
            
            return model, {'clean_accuracy': clean_score}
        
        except Exception as e:
            logger.error(f"Adversarial training failed: {e}")
            # Fallback to regular model
            model = RandomForestClassifier(n_estimators=200, random_state=Config.RANDOM_STATE)
            model.fit(X_train, y_train)
            return model, {'clean_accuracy': model.score(X_test, y_test)}
    
    def detect_evasion_attempt(self, features: Dict, model_predictions: Dict) -> Dict:
        """Detect evasion attempts"""
        evasion_score = 0
        reasons = []
        
        try:
            # Check feature variance
            feature_values = [v for v in features.values() if isinstance(v, (int, float))]
            if feature_values and len(feature_values) > 1:
                variance = np.var(feature_values)
                if variance > 15:
                    evasion_score += 15
                    reasons.append("High feature variance")
            
            # Check model disagreement
            if len(model_predictions) > 1:
                pred_values = [v for v in model_predictions.values() if v != 0]
                if pred_values and len(pred_values) > 1:
                    pred_variance = np.var(pred_values)
                    if pred_variance > 0.2:
                        evasion_score += 20
                        reasons.append("Model disagreement")
            
            is_evasion = evasion_score >= (Config.EVASION_THRESHOLD * 100)
            if is_evasion:
                self.evasion_attempts_detected += 1
            
            return {
                'is_evasion': is_evasion,
                'evasion_score': evasion_score,
                'reasons': reasons
            }
        
        except Exception as e:
            logger.warning(f"Evasion detection error: {e}")
            return {'is_evasion': False, 'evasion_score': 0, 'reasons': []}
    
    def train_evasion_detector(self, X_train: pd.DataFrame, y_train: np.ndarray,
                              X_test: pd.DataFrame, y_test: np.ndarray) -> Dict:
        """Train isolation forest for evasion detection"""
        try:
            X_legitimate = X_train[y_train == 0]
            
            detector = IsolationForest(
                n_estimators=150,
                contamination=0.1,
                random_state=Config.RANDOM_STATE,
                n_jobs=-1
            )
            detector.fit(X_legitimate)
            
            return {'if': detector}
        
        except Exception as e:
            logger.error(f"Evasion detector training failed: {e}")
            return {}


class ThreatIntelligenceFusion:
    """Threat intelligence feeds integration"""
    
    def __init__(self):
        self.openphish_urls = set()
        self.urlhaus_urls = set()
        self.last_update = None
        self.load_threat_feeds()
    
    def load_threat_feeds(self):
        """Load threat intelligence feeds"""
        Colors.print_info("Loading threat intelligence feeds...")
        
        # OpenPhish
        try:
            response = requests.get("https://openphish.com/feed.txt", timeout=10)
            if response.status_code == 200:
                urls = response.text.strip().split('\n')
                self.openphish_urls = set([url.strip() for url in urls if url.strip()][:1000])
                Colors.print_success(f"✓ OpenPhish: {len(self.openphish_urls)} threats")
        except Exception as e:
            Colors.print_warning(f"OpenPhish load failed: {str(e)[:50]}")
            self.openphish_urls = set()
        
        # URLhaus
        try:
            response = requests.get("https://urlhaus.abuse.ch/downloads/csv_recent/", timeout=10)
            if response.status_code == 200:
                lines = response.text.strip().split('\n')[9:]  # Skip header
                urls = []
                for line in lines:
                    try:
                        parts = line.split('","')
                        if len(parts) > 2:
                            urls.append(parts[2].strip('"'))
                    except:
                        continue
                self.urlhaus_urls = set(urls[:1000])
                Colors.print_success(f"✓ URLhaus: {len(self.urlhaus_urls)} threats")
        except Exception as e:
            Colors.print_warning(f"URLhaus load failed: {str(e)[:50]}")
            self.urlhaus_urls = set()
        
        self.last_update = datetime.now()
    
    def check_url_reputation(self, url: str) -> Dict:
        """Check URL against threat feeds"""
        threat_score = 0
        matches = []
        
        try:
            url_lower = url.lower().strip()
            
            # Check OpenPhish
            if any(url_lower in t.lower() or t.lower() in url_lower 
                   for t in self.openphish_urls):
                threat_score = 80
                matches.append('OpenPhish')
            
            # Check URLhaus
            if any(url_lower in t.lower() or t.lower() in url_lower 
                   for t in self.urlhaus_urls):
                threat_score = max(threat_score, 75)
                matches.append('URLhaus')
        
        except Exception as e:
            logger.warning(f"Reputation check error: {e}")
        
        return {
            'combined_risk_score': threat_score,
            'threats_found': matches
        }


class MultiModelEnsemble:
    """Multi-model ensemble system"""
    
    def __init__(self):
        self.models = {}
        self.scalers = {}
        self.feature_names = []
        self.adversarial_defense = AdversarialDefense()
    
    def train_ensemble(self, X_train: np.ndarray, y_train: np.ndarray,
                      X_test: np.ndarray, y_test: np.ndarray) -> Dict:
        """Train ensemble of models"""
        Colors.print_info("Training ensemble models...")
        
        # Random Forest
        self.models['rf'] = RandomForestClassifier(
            n_estimators=200,
            max_depth=20,
            min_samples_split=5,
            min_samples_leaf=2,
            random_state=Config.RANDOM_STATE,
            n_jobs=-1
        )
        self.models['rf'].fit(X_train, y_train)
        rf_score = self.models['rf'].score(X_test, y_test)
        Colors.print_success(f"✓ Random Forest: {rf_score:.2%} accuracy")
        
        # Gradient Boosting
        self.models['gb'] = GradientBoostingClassifier(
            n_estimators=200,
            learning_rate=0.1,
            max_depth=5,
            random_state=Config.RANDOM_STATE
        )
        self.models['gb'].fit(X_train, y_train)
        gb_score = self.models['gb'].score(X_test, y_test)
        Colors.print_success(f"✓ Gradient Boosting: {gb_score:.2%} accuracy")
        
        # Adversarial Robust Model
        if Config.ADVERSARIAL_ENABLED:
            X_train_df = pd.DataFrame(X_train, columns=self.feature_names)
            X_test_df = pd.DataFrame(X_test, columns=self.feature_names)
            
            self.models['adversarial_rf'], adv_results = \
                self.adversarial_defense.train_adversarial_robust_model(
                    X_train_df, y_train, X_test_df, y_test
                )
            Colors.print_success(f"✓ Adversarial Model: {adv_results['clean_accuracy']:.2%} accuracy")
        
        # Evasion Detector
        X_train_df = pd.DataFrame(X_train, columns=self.feature_names)
        X_test_df = pd.DataFrame(X_test, columns=self.feature_names)
        self.models['evasion_detector'] = self.adversarial_defense.train_evasion_detector(
            X_train_df, y_train, X_test_df, y_test
        )
        
        return {'rf_score': rf_score, 'gb_score': gb_score}
    
    def predict_ensemble(self, features: Dict, url: str,
                        threat_intel_score: float = 0,
                        threat_dna_score: float = 0) -> Dict:
        """Make ensemble prediction"""
        try:
            X = pd.DataFrame([features]).reindex(columns=self.feature_names, fill_value=0)
            X_scaled = self.scalers['standard'].transform(X)
            
            predictions = {}
            
            # Get predictions from all models
            if 'rf' in self.models:
                predictions['rf'] = self.models['rf'].predict_proba(X_scaled)[0][1]
            
            if 'gb' in self.models:
                predictions['gb'] = self.models['gb'].predict_proba(X_scaled)[0][1]
            
            if 'adversarial_rf' in self.models:
                predictions['adversarial_rf'] = self.models['adversarial_rf'].predict_proba(X_scaled)[0][1]
            
            # Add external scores
            predictions['threat_intel'] = threat_intel_score / 100.0
            predictions['threat_dna'] = threat_dna_score / 100.0
            
            # Evasion detection
            evasion_result = self.adversarial_defense.detect_evasion_attempt(features, predictions)
            predictions['evasion_detector'] = evasion_result['evasion_score'] / 100.0
            
            # Weighted ensemble
            total_weight = sum(Config.WEIGHTS.get(k, 0.1) for k in predictions)
            weighted_sum = sum(p * Config.WEIGHTS.get(k, 0.1) for k, p in predictions.items())
            final_prob = weighted_sum / total_weight if total_weight > 0 else 0.5
            
            # Boost if evasion detected
            if evasion_result['is_evasion']:
                final_prob = min(final_prob * 1.3, 1.0)
            
            return {
                'final_probability': final_prob,
                'is_phishing': final_prob > 0.5,
                'model_predictions': predictions,
                'confidence': max(final_prob, 1 - final_prob),
                'evasion_detection': evasion_result
            }
        
        except Exception as e:
            logger.error(f"Ensemble prediction error: {e}")
            return {
                'final_probability': 0.5,
                'is_phishing': False,
                'model_predictions': {},
                'confidence': 0.5,
                'evasion_detection': {'is_evasion': False, 'evasion_score': 0, 'reasons': []}
            }


# ==================== FEATURE EXTRACTION ====================

def extract_features(url: str) -> Tuple[Optional[Dict], Optional[str]]:
    """Extract features from URL with comprehensive analysis"""
    try:
        # Normalize URL
        if not url.startswith(('http://', 'https://')):
            url = 'https://' + url
        
        parsed = urlparse(url)
        domain = parsed.netloc.lower()
        path = parsed.path.lower()
        url_lower = url.lower()
        
        # Suspicious keywords
        suspicious_keywords = [
            'verify', 'account', 'update', 'confirm', 'login', 'secure',
            'banking', 'paypal', 'amazon', 'netflix', 'suspended', 'locked',
            'verify', 'signin', 'webscr', 'password', 'credential'
        ]
        
        # Suspicious TLDs
        suspicious_tlds = [
            '.xyz', '.top', '.win', '.club', '.online', '.site', '.tk',
            '.ml', '.ga', '.cf', '.gq', '.pw', '.cc'
        ]
        
        # Brand names
        brand_names = [
            'paypal', 'amazon', 'netflix', 'linkedin', 'facebook', 
            'google', 'microsoft', 'apple', 'ebay', 'instagram'
        ]
        
        # Calculate features
        keyword_count = sum(1 for kw in suspicious_keywords if kw in url_lower)
        has_suspicious_tld = any(tld in domain for tld in suspicious_tlds)
        has_brand_name = any(brand in domain for brand in brand_names)
        has_typo = any(char in domain for char in ['1', '0', 'l']) and has_brand_name
        has_executable = any(ext in path for ext in ['.exe', '.scr', '.bat', '.cmd', '.zip'])
        
        features = {
            'url_length': min(len(url) / 100.0, 2.0),
            'domain_length': min(len(domain) / 50.0, 2.0),
            'subdomain_count': min(max(0, domain.count('.') - 1), 5),
            'has_https': 1 if url.startswith('https://') else 0,
            'has_ip': 2 if re.match(r'(\d+\.){3}\d+', domain) else 0,
            'login_word': 2 if 'login' in url_lower else 0,
            'entropy': min(calculate_entropy(url) / 5.0, 2.0),
            'keyword_count': min(keyword_count, 5),
            'has_suspicious_tld': 3 if has_suspicious_tld else 0,
            'has_brand_typo': 3 if has_typo else 0,
            'has_executable': 5 if has_executable else 0,
            'multiple_subdomains': 2 if len(domain.split('.')) > 3 else 0,
            'has_hyphen': 1 if '-' in domain else 0,
            'has_numbers': 1 if any(c.isdigit() for c in domain) else 0,
            'has_at_symbol': 3 if '@' in url else 0,
            'suspicious_pattern': 2 if (has_brand_name and not domain.endswith(('.com', '.org', '.net'))) else 0
        }
        
        return features, url
    
    except Exception as e:
        logger.error(f"Feature extraction failed: {e}")
        return None, None


def calculate_entropy(text: str) -> float:
    """Calculate Shannon entropy of text"""
    if not text:
        return 0
    
    entropy = 0
    for char in set(text):
        p = text.count(char) / len(text)
        entropy -= p * np.log2(p)
    
    return entropy


def load_data(file_path: str) -> pd.DataFrame:
    """Load and preprocess dataset"""
    try:
        Colors.print_info(f"Loading dataset from: {file_path}")
        df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        
        # Find URL and label columns
        url_col = label_col = None
        for col in df.columns:
            col_lower = col.lower()
            if col_lower in ['url', 'text', 'link', 'website']:
                url_col = col
            if col_lower in ['label', 'class', 'target', 'type']:
                label_col = col
        
        if url_col is None or label_col is None:
            raise ValueError("Could not find URL or label columns")
        
        # Clean data
        df_clean = df[[url_col, label_col]].copy()
        df_clean.columns = ['url', 'label']
        df_clean = df_clean.dropna()
        df_clean['label'] = pd.to_numeric(df_clean['label'], errors='coerce').astype(int)
        df_clean = df_clean[df_clean['label'].isin([0, 1])]
        
        Colors.print_success(f"✓ Loaded {len(df_clean)} records")
        Colors.print_info(f"  Phishing: {sum(df_clean['label'] == 1)} | Legitimate: {sum(df_clean['label'] == 0)}")
        
        return df_clean
    
    except Exception as e:
        logger.error(f"Data loading failed: {e}")
        raise


def train_model(df: pd.DataFrame) -> Dict:
    """Train the complete model pipeline"""
    Colors.print_header("TRAINING MODEL PIPELINE")
    
    # Extract features
    Colors.print_info("Extracting features from URLs...")
    features_list = []
    labels = []
    
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Processing URLs"):
        feat, _ = extract_features(row['url'])
        if feat:
            features_list.append(feat)
            labels.append(row['label'])
    
    X = pd.DataFrame(features_list).fillna(0)
    y = np.array(labels)
    
    Colors.print_success(f"✓ Extracted {len(X)} feature vectors")
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=Config.TEST_SIZE, random_state=Config.RANDOM_STATE, stratify=y
    )
    
    Colors.print_info(f"Training set: {len(X_train)} | Test set: {len(X_test)}")
    
    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Train ensemble
    ensemble = MultiModelEnsemble()
    ensemble.feature_names = list(X.columns)
    ensemble.scalers['standard'] = scaler
    
    ensemble.train_ensemble(X_train_scaled, y_train, X_test_scaled, y_test)
    
    # Evaluate
    Colors.print_info("Evaluating ensemble performance...")
    test_preds = []
    for i in range(len(X_test)):
        result = ensemble.predict_ensemble(X_test.iloc[i].to_dict(), '', 0, 0)
        test_preds.append(result['is_phishing'])
    
    accuracy = accuracy_score(y_test, test_preds)
    
    Colors.print_success(f"✓ Final Ensemble Accuracy: {accuracy:.2%}")
    
    # Save model
    model_data = {
        'ensemble': ensemble,
        'scaler': scaler,
        'feature_names': list(X.columns),
        'accuracy': accuracy
    }
    
    import joblib
    joblib.dump(model_data, Config.MODEL_FILE)
    Colors.print_success(f"✓ Model saved to: {Config.MODEL_FILE}")
    
    return model_data


def predict_url_with_defenses(url: str, model_data: Dict, 
                              threat_intel: ThreatIntelligenceFusion,
                              threat_dna_score: float = 0) -> Dict:
    """Predict URL with all defense mechanisms"""
    features, normalized_url = extract_features(url)
    
    if not features:
        return {'error': 'Feature extraction failed'}
    
    try:
        # Get predictions
        ensemble = model_data['ensemble']
        threat_result = threat_intel.check_url_reputation(normalized_url)
        threat_score = threat_result['combined_risk_score']
        
        ensemble_result = ensemble.predict_ensemble(
            features, normalized_url, threat_score, threat_dna_score
        )
        
        ml_risk = int(ensemble_result['final_probability'] * 100)
        
        # Heuristic analysis
        heuristic_score = 0
        url_lower = normalized_url.lower()
        
        if features.get('has_executable', 0) > 0:
            heuristic_score = max(heuristic_score, 85)
        if features.get('has_brand_typo', 0) > 0:
            heuristic_score = max(heuristic_score, 80)
        if features.get('has_suspicious_tld', 0) > 0:
            heuristic_score = max(heuristic_score, 75)
        
        keyword_count = features.get('keyword_count', 0)
        if keyword_count >= 4:
            heuristic_score = max(heuristic_score, 75)
        elif keyword_count >= 3:
            heuristic_score = max(heuristic_score, 65)
        elif keyword_count >= 2:
            heuristic_score = max(heuristic_score, 55)
        
        if 'verify' in url_lower and 'account' in url_lower:
            heuristic_score = max(heuristic_score, 70)
        
        # Calculate final risk
        if heuristic_score >= 75:
            base_risk = int(ml_risk * 0.15 + heuristic_score * 0.85)
        elif heuristic_score >= 60:
            base_risk = int(ml_risk * 0.25 + heuristic_score * 0.75)
        elif heuristic_score > 0:
            base_risk = int(ml_risk * 0.5 + heuristic_score * 0.5)
        else:
            base_risk = ml_risk
        
        # Incorporate threat intelligence
        if threat_score > 0:
            base_risk = min(int(base_risk * 0.6 + threat_score * 0.4), 100)
        
        base_risk = min(base_risk, 100)
        
        # Determine risk level
        if base_risk >= Config.HIGH_RISK:
            risk_level = "HIGH RISK"
            recommendation = "⛔ DO NOT OPEN - High probability of phishing"
            is_phishing = True
        elif base_risk >= Config.MEDIUM_RISK:
            risk_level = "MEDIUM RISK"
            recommendation = "⚠️ VERIFY FIRST - Exercise caution"
            is_phishing = True
        else:
            risk_level = "SAFE"
            recommendation = "✅ Appears legitimate"
            is_phishing = False
        
        return {
            'url': normalized_url,
            'is_phishing': is_phishing,
            'risk_score': base_risk,
            'risk_level': risk_level,
            'recommendation': recommendation,
            'confidence': ensemble_result['confidence'],
            'model_predictions': ensemble_result['model_predictions'],
            'evasion_detection': ensemble_result['evasion_detection'],
            'threat_intelligence': threat_result,
            'heuristic_score': heuristic_score,
            'ml_risk': ml_risk
        }
    
    except Exception as e:
        logger.error(f"Prediction error: {e}")
        return {'error': f'Prediction failed: {str(e)}'}


def display_result(result: Dict, show_visualization: bool = True):
    """Display analysis results with beautiful formatting and optional visualization"""
    Colors.print_header("🛡️ ANALYSIS RESULT")
    
    print(f"\n{Colors.BOLD}URL:{Colors.ENDC} {result['url'][:70]}...")
    
    # Status
    status_color = Colors.FAIL if result['is_phishing'] else Colors.OKGREEN
    status = "⚠️ PHISHING DETECTED" if result['is_phishing'] else "✅ LEGITIMATE"
    print(f"\n{Colors.BOLD}Status:{Colors.ENDC} {status_color}{status}{Colors.ENDC}")
    
    # Risk bar
    risk = result['risk_score']
    bar_length = 50
    filled = int(bar_length * risk / 100)
    bar_color = Colors.FAIL if risk >= 70 else Colors.WARNING if risk >= 45 else Colors.OKGREEN
    bar = '█' * filled + '░' * (bar_length - filled)
    print(f"\n{Colors.BOLD}Risk Score:{Colors.ENDC} {bar_color}{bar}{Colors.ENDC} {risk}/100")
    print(f"{Colors.BOLD}Risk Level:{Colors.ENDC} {result['risk_level']}")
    print(f"{Colors.BOLD}Recommendation:{Colors.ENDC} {result['recommendation']}")
    
    # Additional info
    if result.get('threat_intelligence', {}).get('threats_found'):
        threats = ', '.join(result['threat_intelligence']['threats_found'])
        print(f"\n{Colors.FAIL}⚠️ Matched threat feeds: {threats}{Colors.ENDC}")
    
    if result.get('evasion_detection', {}).get('is_evasion'):
        print(f"{Colors.WARNING}⚠️ Evasion attempt detected!{Colors.ENDC}")
    
    # Generate visualization
    if show_visualization and risk >= 0:
        try:
            viz_path = generate_url_visualization(result)
            if viz_path:
                Colors.print_success(f"\n✓ Visualization saved: {os.path.basename(viz_path)}")
                Colors.print_info("✓ Opening visualization in browser...")
                try:
                    webbrowser.open(f'file://{os.path.abspath(viz_path)}')
                    Colors.print_success("✓ Visualization opened in browser")
                except Exception as e:
                    Colors.print_warning(f"Could not auto-open visualization: {e}")
        except Exception as e:
            logger.warning(f"Visualization generation failed: {e}")
    
    # Auto-open URL if safe
    if not result['is_phishing'] and risk < Config.MEDIUM_RISK:
        try:
            Colors.print_success(f"\n✓ URL is safe - Opening in browser...")
            webbrowser.open(result['url'])
            Colors.print_success("✓ URL opened successfully")
        except Exception as e:
            Colors.print_error(f"Failed to open URL: {e}")
    elif result['is_phishing']:
        print(f"\n{Colors.FAIL}⛔ URL NOT opened - Phishing detected!{Colors.ENDC}")
    
    print()


def generate_url_visualization(result: Dict) -> Optional[str]:
    """Generate visualization for URL analysis"""
    try:
        fig, axes = plt.subplots(2, 2, figsize=(14, 10))
        
        # Risk gauge
        ax1 = axes[0, 0]
        risk = result['risk_score']
        colors = ['green', 'yellow', 'orange', 'red']
        segments = [Config.MEDIUM_RISK, Config.HIGH_RISK - Config.MEDIUM_RISK, 
                   100 - Config.HIGH_RISK]
        
        wedges, texts = ax1.pie([Config.MEDIUM_RISK, Config.HIGH_RISK - Config.MEDIUM_RISK, 
                                 100 - Config.HIGH_RISK],
                                colors=['green', 'orange', 'red'],
                                startangle=180, counterclock=False)
        
        # Add needle
        angle = 180 - (risk * 1.8)
        ax1.annotate('', xy=(0.6 * np.cos(np.radians(angle)), 
                             0.6 * np.sin(np.radians(angle))),
                    xytext=(0, 0),
                    arrowprops=dict(arrowstyle='->', lw=3, color='black'))
        
        ax1.set_title(f'Risk Score: {risk}/100', fontweight='bold', fontsize=12)
        
        # Model predictions
        ax2 = axes[0, 1]
        if result.get('model_predictions'):
            preds = result['model_predictions']
            names = list(preds.keys())
            values = [v * 100 for v in preds.values()]
            colors_bar = ['green' if v < 50 else 'orange' if v < 70 else 'red' for v in values]
            
            ax2.barh(names, values, color=colors_bar, alpha=0.7)
            ax2.set_xlabel('Prediction Confidence (%)')
            ax2.set_title('Model Predictions', fontweight='bold')
            ax2.axvline(x=50, color='black', linestyle='--', alpha=0.5)
            ax2.grid(axis='x', alpha=0.3)
        
        # Feature breakdown
        ax3 = axes[1, 0]
        feature_scores = {
            'Heuristic': result.get('heuristic_score', 0),
            'ML Model': result.get('ml_risk', 0),
            'Threat Intel': result.get('threat_intelligence', {}).get('combined_risk_score', 0),
        }
        
        names = list(feature_scores.keys())
        values = list(feature_scores.values())
        colors_feat = ['green' if v < 45 else 'orange' if v < 70 else 'red' for v in values]
        
        ax3.bar(names, values, color=colors_feat, alpha=0.7)
        ax3.set_ylabel('Risk Score')
        ax3.set_title('Risk Components', fontweight='bold')
        ax3.axhline(y=Config.HIGH_RISK, color='red', linestyle='--', alpha=0.5, label='High Risk')
        ax3.axhline(y=Config.MEDIUM_RISK, color='orange', linestyle='--', alpha=0.5, label='Medium Risk')
        ax3.legend()
        ax3.grid(axis='y', alpha=0.3)
        
        # Summary text
        ax4 = axes[1, 1]
        ax4.axis('off')
        
        summary_text = f"""
URL ANALYSIS SUMMARY

URL: {result['url'][:60]}...

Status: {result['risk_level']}
Risk Score: {risk}/100
Confidence: {result.get('confidence', 0):.1%}

Recommendation:
{result['recommendation']}

Analysis Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
        """
        
        ax4.text(0.1, 0.5, summary_text, fontsize=10, verticalalignment='center',
                fontfamily='monospace',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.3))
        
        plt.suptitle(f'🛡️ PHISHING ANALYSIS - {result["risk_level"]}',
                    fontsize=14, fontweight='bold')
        plt.tight_layout()
        
        viz_file = f'url_analysis_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png'
        viz_path = os.path.join(Config.VISUALIZATION_DIR, viz_file)
        plt.savefig(viz_path, dpi=Config.VIZ_DPI, bbox_inches='tight')
        plt.close()
        
        return viz_path
    
    except Exception as e:
        logger.error(f"Visualization generation failed: {e}")
        return None


# ==================== MAIN APPLICATION ====================

def main():
    """Main application entry point"""
    Colors.print_header("🧬 PHISHING DETECTION SYSTEM v10.0 - IMPROVED")
    print(f"{Colors.OKCYAN}Enhanced: Bug Fixes | Better Crawling | Improved Detection{Colors.ENDC}\n")
    
    # Check configuration
    if not Config.DATA_FILE:
        Colors.print_error("Dataset not found!")
        print("\nPlease place your dataset in one of these locations:")
        for path in Config.POSSIBLE_DATA_PATHS:
            print(f"  • {path}")
        print("\nOr update Config.DATA_FILE with your dataset path.")
        return
    
    # Initialize systems
    Colors.print_info("Initializing systems...")
    threat_intel = ThreatIntelligenceFusion()
    monitor = RealTimeMonitor()
    cyber_map = CyberMapVisualizer()
    chat_assistant = SecurityChatAssistant()
    
    # Advanced systems
    crawler = ThreatDNACrawler() if HAS_PLAYWRIGHT else None
    embedding_gen = EmbeddingGenerator() if HAS_TRANSFORMERS else None
    vector_db = ThreatVectorDB() if HAS_VECTORDB else None
    campaign_detector = CampaignDetector()
    
    # Show capabilities
    print(f"\n{Colors.BOLD}System Capabilities:{Colors.ENDC}")
    print(f"  Playwright:    {'✓ Enabled' if HAS_PLAYWRIGHT else '✗ Disabled'} - Web Crawling")
    print(f"  Transformers:  {'✓ Enabled' if HAS_TRANSFORMERS else '✗ Disabled'} - Embeddings")
    print(f"  VectorDB:      {'✓ Enabled' if HAS_VECTORDB else '✗ Disabled'} - Similarity Search")
    print(f"  OpenCV:        {'✓ Enabled' if HAS_CV else '✗ Disabled'} - Image Analysis")
    print(f"  NetworkX:      {'✓ Enabled' if HAS_NETWORKX else '✗ Disabled'} - Graph Analysis")
    
    try:
        # Load or train model
        if os.path.exists(Config.MODEL_FILE):
            Colors.print_info("\nLoading trained model...")
            import joblib
            model_data = joblib.load(Config.MODEL_FILE)
            Colors.print_success(f"✓ Model loaded (Accuracy: {model_data['accuracy']:.2%})")
        else:
            Colors.print_warning("\nNo trained model found. Training new model...")
            df = load_data(Config.DATA_FILE)
            model_data = train_model(df)
        
        # Start monitoring
        monitor.start_monitoring()
        
        # Collected DNA samples
        collected_dnas = []
        
        # Main menu loop
        while True:
            print(f"\n{Colors.BOLD}{'='*80}")
            print(f"  🧬 PHISHING DETECTION v10.0 - COMMAND CENTER")
            print(f"{'='*80}{Colors.ENDC}")
            
            print(f"\n{Colors.HEADER}═══ CORE ANALYSIS ═══{Colors.ENDC}")
            print(f"{Colors.OKCYAN}[1]{Colors.ENDC}  🔗 Quick URL Scan")
            print(f"{Colors.OKCYAN}[2]{Colors.ENDC}  🛡️ Full Analysis with Report")
            
            print(f"\n{Colors.HEADER}═══ 🧬 THREAT DNA ═══{Colors.ENDC}")
            print(f"{Colors.OKGREEN}[14]{Colors.ENDC} 🕷️  Deep Crawl & DNA Collection")
            print(f"{Colors.OKGREEN}[15]{Colors.ENDC} 🔍 Find Similar Threats")
            print(f"{Colors.OKGREEN}[16]{Colors.ENDC} 🎯 Campaign Detection")
            print(f"{Colors.OKGREEN}[17]{Colors.ENDC} 📊 DNA Database Stats")
            
            print(f"\n{Colors.HEADER}═══ ADVANCED FEATURES ═══{Colors.ENDC}")
            print(f"{Colors.OKGREEN}[10]{Colors.ENDC} 🌍 3D Cyber Map")
            print(f"{Colors.OKGREEN}[11]{Colors.ENDC} 📊 Real-Time Dashboard")
            print(f"{Colors.OKGREEN}[12]{Colors.ENDC} 🤖 AI Chat Assistant")
            print(f"{Colors.OKGREEN}[13]{Colors.ENDC} 📈 Analytics & Trends")
            
            print(f"\n{Colors.HEADER}═══ UTILITIES ═══{Colors.ENDC}")
            print(f"{Colors.OKCYAN}[4]{Colors.ENDC}  📊 Threat Intel Stats")
            print(f"{Colors.OKCYAN}[5]{Colors.ENDC}  🔄 Reload Threat Feeds")
            print(f"{Colors.OKCYAN}[8]{Colors.ENDC}  ❓ Help & Information")
            print(f"{Colors.OKCYAN}[9]{Colors.ENDC}  🚪 Exit")
            print(f"{Colors.BOLD}{'='*80}{Colors.ENDC}")
            
            choice = input(f"\n{Colors.BOLD}Select option:{Colors.ENDC} ").strip()
            
            if choice == '9':
                monitor.stop_monitoring()
                Colors.print_success("\n👋 Stay safe! Goodbye!")
                break
            
            elif choice == '1':
                url = input("\n🔗 Enter URL: ").strip()
                if not url:
                    continue
                
                Colors.print_info("Analyzing URL...")
                result = predict_url_with_defenses(url, model_data, threat_intel)
                if result.get('error'):
                    Colors.print_error(result['error'])
                else:
                    monitor.log_detection(result)
                    display_result(result, show_visualization=True)
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '2':
                url = input("\n🔗 Enter URL: ").strip()
                if not url:
                    continue
                
                Colors.print_info("Running full analysis...")
                result = predict_url_with_defenses(url, model_data, threat_intel)
                if result.get('error'):
                    Colors.print_error(result['error'])
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                monitor.log_detection(result)
                display_result(result, show_visualization=True)
                
                save = input(f"\n{Colors.BOLD}Save detailed JSON report? (y/n):{Colors.ENDC} ").strip().lower()
                if save == 'y':
                    report_file = f"report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
                    report_path = os.path.join(Config.REPORTS_DIR, report_file)
                    with open(report_path, 'w') as f:
                        json.dump(result, f, indent=2, default=str)
                    Colors.print_success(f"✓ Report saved: {report_file}")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '4':
                Colors.print_header("📊 THREAT INTELLIGENCE STATISTICS")
                print(f"\n  OpenPhish URLs:    {len(threat_intel.openphish_urls)}")
                print(f"  URLhaus URLs:      {len(threat_intel.urlhaus_urls)}")
                if threat_intel.last_update:
                    print(f"  Last Update:       {threat_intel.last_update.strftime('%Y-%m-%d %H:%M:%S')}")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '5':
                Colors.print_info("Reloading threat intelligence feeds...")
                threat_intel.load_threat_feeds()
                Colors.print_success("✓ Feeds reloaded successfully!")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '8':
                Colors.print_header("❓ HELP & INFORMATION")
                print(f"\n{Colors.BOLD}🧬 THREAT DNA:{Colors.ENDC}")
                print("  • Deep web crawling in isolated browser")
                print("  • Multi-modal embeddings (text, image, DOM)")
                print("  • Vector similarity search for threat correlation")
                print("  • Campaign detection using ML clustering\n")
                
                print(f"{Colors.BOLD}🛡️ ADVERSARIAL DEFENSE:{Colors.ENDC}")
                print("  • Adversarial training for robustness")
                print("  • Evasion attempt detection")
                print("  • Multi-model ensemble voting\n")
                
                print(f"{Colors.BOLD}🌐 THREAT INTELLIGENCE:{Colors.ENDC}")
                print("  • OpenPhish live feed integration")
                print("  • URLhaus malware database")
                print("  • Real-time threat correlation\n")
                
                print(f"{Colors.BOLD}📊 FEATURES:{Colors.ENDC}")
                print("  • 16+ URL features extracted")
                print("  • Heuristic + ML hybrid analysis")
                print("  • Real-time monitoring & dashboards")
                print("  • 3D visualization & mapping\n")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '10':
                Colors.print_header("🌍 3D CYBER MAP GENERATOR")
                if not monitor.detection_history:
                    Colors.print_warning("No data available. Analyze some URLs first!")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                Colors.print_info("Generating 3D cyber threat map...")
                map_path = cyber_map.create_3d_cyber_map(list(monitor.detection_history))
                
                if map_path:
                    Colors.print_success(f"✓ Map saved: {os.path.basename(map_path)}")
                    Colors.print_info(f"✓ Opening map in browser...")
                    try:
                        webbrowser.open(f'file://{os.path.abspath(map_path)}')
                        Colors.print_success("✓ Map opened successfully in browser")
                    except Exception as e:
                        Colors.print_error(f"Failed to open map: {e}")
                        Colors.print_info(f"Manual path: {map_path}")
                else:
                    Colors.print_error("Failed to generate map")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '11':
                Colors.print_header("📊 REAL-TIME DASHBOARD")
                stats = monitor.get_current_stats()
                
                print(f"\n{Colors.BOLD}LIVE STATISTICS{Colors.ENDC}")
                print(f"  Total Scans:       {stats['total_scans']}")
                print(f"  Phishing Detected: {stats['phishing_detected']}")
                print(f"  Safe URLs:         {stats['safe_urls']}")
                print(f"  Detection Rate:    {stats['detection_rate']:.1f}%")
                
                print(f"\n{Colors.BOLD}SYSTEM HEALTH{Colors.ENDC}")
                print(f"  CPU Usage:         {stats['cpu_usage']:.1f}%")
                print(f"  Memory Usage:      {stats['memory_usage']:.1f}%")
                print(f"  Active Threats:    {stats['active_threats']}")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '12':
                Colors.print_header("🤖 AI SECURITY CHAT ASSISTANT")
                print(f"\n{Colors.OKCYAN}Welcome! I'm your AI Security Copilot.{Colors.ENDC}")
                print(f"{Colors.BOLD}Type 'back' to return to main menu{Colors.ENDC}\n")
                
                # Update context
                chat_assistant.update_context(
                    detections=list(monitor.detection_history),
                    system_health={
                        'cpu': monitor.system_metrics[-1]['cpu_percent'] if monitor.system_metrics else 0,
                        'memory': monitor.system_metrics[-1]['memory_percent'] if monitor.system_metrics else 0
                    }
                )
                
                while True:
                    user_input = input(f"\n{Colors.OKBLUE}You:{Colors.ENDC} ").strip()
                    if user_input.lower() == 'back':
                        break
                    if not user_input:
                        continue
                    
                    response = chat_assistant.process_query(user_input)
                    print(f"\n{Colors.OKGREEN}Assistant:{Colors.ENDC} {response}")
            
            elif choice == '13':
                Colors.print_header("📈 ADVANCED ANALYTICS")
                if not monitor.detection_history:
                    Colors.print_warning("No data available. Analyze some URLs first!")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                detections = list(monitor.detection_history)
                total = len(detections)
                phishing = sum(1 for d in detections if d['is_phishing'])
                
                print(f"\n{Colors.BOLD}DETECTION TRENDS{Colors.ENDC}")
                print(f"  Total Analyzed:     {total}")
                print(f"  Phishing Found:     {phishing} ({phishing/total*100:.1f}%)")
                print(f"  Safe URLs:          {total - phishing}")
                
                high_risk = sum(1 for d in detections if d['risk_score'] >= Config.HIGH_RISK)
                medium_risk = sum(1 for d in detections if Config.MEDIUM_RISK <= d['risk_score'] < Config.HIGH_RISK)
                low_risk = total - high_risk - medium_risk
                
                print(f"\n{Colors.BOLD}RISK DISTRIBUTION{Colors.ENDC}")
                print(f"  {Colors.FAIL}High Risk:    {high_risk:3d} ({high_risk/total*100:5.1f}%){Colors.ENDC}")
                print(f"  {Colors.WARNING}Medium Risk:  {medium_risk:3d} ({medium_risk/total*100:5.1f}%){Colors.ENDC}")
                print(f"  {Colors.OKGREEN}Low Risk:     {low_risk:3d} ({low_risk/total*100:5.1f}%){Colors.ENDC}")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '14':
                Colors.print_header("🕷️ DEEP CRAWL & DNA COLLECTION")
                
                if not crawler:
                    Colors.print_error("Playwright not installed!")
                    print(f"\nInstall with: pip install playwright && playwright install chromium")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                url = input("\n🔗 Enter URL to crawl: ").strip()
                if not url:
                    continue
                
                Colors.print_info("🕷️ Starting deep crawl (may take 30-60 seconds)...")
                print(f"{Colors.WARNING}⚠️  URL will be opened in isolated browser environment{Colors.ENDC}\n")
                
                # Crawl URL
                dna = crawler.crawl_url_sync(url)
                
                if not dna:
                    Colors.print_warning("Crawl failed - URL may be unreachable")
                    Colors.print_info("Running standard analysis instead...")
                    
                    # Fallback to standard analysis
                    features, _ = extract_features(url)
                    if features:
                        result = predict_url_with_defenses(url, model_data, threat_intel)
                        if not result.get('error'):
                            monitor.log_detection(result)
                            display_result(result)
                    
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                # Enrich with embeddings
                if embedding_gen:
                    Colors.print_info("🔢 Generating multi-modal embeddings...")
                    dna = embedding_gen.enrich_threat_dna(dna)
                    if dna.combined_embedding is not None:
                        Colors.print_success("✓ Embeddings generated")
                    else:
                        Colors.print_warning("⚠️ Embedding generation incomplete")
                
                # Run standard analysis
                features, _ = extract_features(url)
                if features:
                    result = predict_url_with_defenses(url, model_data, threat_intel)
                    dna.risk_score = result['risk_score']
                    dna.threat_category = result['risk_level']
                    monitor.log_detection(result)
                    
                    # Generate visualization for DNA analysis
                    try:
                        viz_path = generate_url_visualization(result)
                        if viz_path:
                            Colors.print_success(f"✓ Visualization: {os.path.basename(viz_path)}")
                            Colors.print_info("✓ Opening visualization in browser...")
                            try:
                                webbrowser.open(f'file://{os.path.abspath(viz_path)}')
                                Colors.print_success("✓ Visualization opened")
                            except:
                                pass
                    except:
                        pass
                
                # Store in vector DB
                if vector_db and dna.combined_embedding is not None:
                    Colors.print_info("💾 Storing in vector database...")
                    if vector_db.add_threat(dna):
                        Colors.print_success("✓ Stored in vector DB")
                
                # Save DNA sample
                collected_dnas.append(dna)
                
                # Display results
                Colors.print_header("🧬 THREAT DNA ANALYSIS")
                print(f"\n{Colors.BOLD}Threat ID:{Colors.ENDC} {dna.threat_id}")
                print(f"{Colors.BOLD}URL:{Colors.ENDC} {url[:70]}...")
                print(f"{Colors.BOLD}Risk Score:{Colors.ENDC} {dna.risk_score}/100")
                print(f"{Colors.BOLD}Category:{Colors.ENDC} {dna.threat_category or 'N/A'}")
                
                print(f"\n{Colors.BOLD}━━━ Collected Artifacts ━━━{Colors.ENDC}")
                print(f"  Screenshot:     {'✓' if dna.screenshot_path else '✗'}")
                print(f"  HTML Dump:      {'✓' if dna.html_content else '✗'}")
                print(f"  DOM Snapshot:   {'✓' if dna.dom_snapshot else '✗'}")
                print(f"  Console Logs:   {len(dna.console_logs)} entries")
                print(f"  Network Reqs:   {len(dna.network_requests)} requests")
                
                print(f"\n{Colors.BOLD}━━━ Extracted Features ━━━{Colors.ENDC}")
                print(f"  Page Title:     {dna.page_title or 'N/A'}")
                print(f"  Text Content:   {len(dna.text_content or '')} chars")
                print(f"  Links Found:    {len(dna.extracted_links)}")
                print(f"  Forms Found:    {len(dna.extracted_forms)}")
                if dna.response_time_ms:
                    print(f"  Response Time:  {dna.response_time_ms:.0f}ms")
                
                print(f"\n{Colors.BOLD}━━━ Embeddings Generated ━━━{Colors.ENDC}")
                print(f"  Text Embedding:     {'✓' if dna.text_embedding is not None else '✗'}")
                print(f"  Image Embedding:    {'✓' if dna.image_embedding is not None else '✗'}")
                print(f"  DOM Embedding:      {'✓' if dna.dom_embedding is not None else '✗'}")
                print(f"  Combined Embedding: {'✓' if dna.combined_embedding is not None else '✗'}")
                
                if dna.javascript_behavior:
                    print(f"\n{Colors.BOLD}━━━ JavaScript Behavior ━━━{Colors.ENDC}")
                    for key, value in list(dna.javascript_behavior.items())[:5]:
                        print(f"  {key}: {value}")
                
                # Find similar threats
                if vector_db and dna.combined_embedding is not None:
                    similar = vector_db.find_similar_threats(dna, top_k=3)
                    if similar:
                        print(f"\n{Colors.BOLD}━━━ Similar Threats Found ━━━{Colors.ENDC}")
                        for i, sim in enumerate(similar, 1):
                            print(f"  {i}. Similarity: {sim['similarity']:.1%} | Risk: {sim['metadata'].get('risk_score', 0)}/100")
                            print(f"     URL: {sim['metadata'].get('url', 'N/A')[:60]}...")
                
                print(f"\n{Colors.BOLD}━━━ DNA Fingerprint ━━━{Colors.ENDC}")
                print(f"  {dna.get_fingerprint()}")
                
                # Save DNA to file
                dna_file = f"threat_dna_{dna.threat_id}.json"
                dna_path = os.path.join(Config.THREAT_DNA_DIR, dna_file)
                with open(dna_path, 'w', encoding='utf-8') as f:
                    json.dump(dna.to_dict(), f, indent=2, ensure_ascii=False)
                Colors.print_success(f"\n✓ DNA sample saved: {dna_file}")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '15':
                Colors.print_header("🔍 FIND SIMILAR THREATS (Vector Search)")
                
                if not vector_db:
                    Colors.print_error("Vector DB not available!")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                if not collected_dnas:
                    Colors.print_warning("No DNA samples collected yet. Use option [14] first!")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                print(f"\n{Colors.BOLD}Available DNA Samples:{Colors.ENDC}")
                for i, dna in enumerate(collected_dnas[-10:], 1):
                    print(f"  [{i}] {dna.threat_id} - {dna.url[:50]}... (Risk: {dna.risk_score}/100)")
                
                try:
                    idx = int(input(f"\n{Colors.BOLD}Select sample (1-{min(len(collected_dnas), 10)}):{Colors.ENDC} ").strip())
                    if idx < 1 or idx > min(len(collected_dnas), 10):
                        Colors.print_error("Invalid selection!")
                        input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                        continue
                    
                    # Get selected DNA (accounting for the slice)
                    available_dnas = collected_dnas[-10:]
                    selected_dna = available_dnas[idx - 1]
                    
                    if selected_dna.combined_embedding is None:
                        Colors.print_error("No embedding available for this sample!")
                        input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                        continue
                    
                    Colors.print_info("🔍 Searching for similar threats...")
                    similar = vector_db.find_similar_threats(selected_dna, top_k=5)
                    
                    if not similar:
                        Colors.print_warning("No similar threats found in database")
                    else:
                        print(f"\n{Colors.BOLD}━━━ SIMILAR THREATS FOUND ━━━{Colors.ENDC}\n")
                        for i, sim in enumerate(similar, 1):
                            similarity = sim['similarity']
                            metadata = sim['metadata']
                            
                            color = Colors.FAIL if similarity > 0.9 else Colors.WARNING if similarity > 0.8 else Colors.OKGREEN
                            print(f"{color}[{i}] Similarity: {similarity:.1%}{Colors.ENDC}")
                            print(f"    Threat ID:  {sim['threat_id']}")
                            print(f"    URL:        {metadata.get('url', 'N/A')[:70]}...")
                            print(f"    Risk:       {metadata.get('risk_score', 0)}/100")
                            print(f"    Category:   {metadata.get('threat_category', 'unknown')}")
                            print(f"    Timestamp:  {metadata.get('timestamp', 'N/A')[:19]}\n")
                        
                        if similar and similar[0]['similarity'] > 0.9:
                            print(f"{Colors.FAIL}⚠️ WARNING: Very high similarity detected!")
                            print(f"   Possible campaign or reused infrastructure.{Colors.ENDC}")
                
                except ValueError:
                    Colors.print_error("Invalid input! Please enter a number.")
                except Exception as e:
                    Colors.print_error(f"Error: {e}")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '16':
                Colors.print_header("🎯 CAMPAIGN DETECTION & CLUSTERING")
                
                if not collected_dnas:
                    Colors.print_warning("No DNA samples collected yet. Use option [14] first!")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                if len(collected_dnas) < Config.CAMPAIGN_MIN_SAMPLES:
                    Colors.print_warning(f"Need at least {Config.CAMPAIGN_MIN_SAMPLES} samples for clustering.")
                    Colors.print_info(f"Currently have: {len(collected_dnas)} samples")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
                    continue
                
                Colors.print_info("🎯 Running campaign detection (DBSCAN clustering)...")
                campaigns = campaign_detector.detect_campaigns(collected_dnas)
                
                print(f"\n{Colors.BOLD}━━━ CAMPAIGN DETECTION RESULTS ━━━{Colors.ENDC}\n")
                print(f"Total Campaigns Detected: {len([c for c in campaigns if c != 'UNCLUSTERED'])}")
                print(f"Unclustered Threats: {len(campaigns.get('UNCLUSTERED', []))}\n")
                
                for campaign_id, threats in campaigns.items():
                    if campaign_id == 'UNCLUSTERED':
                        if threats:
                            print(f"\n{Colors.BOLD}UNCLUSTERED ({len(threats)} threats){Colors.ENDC}")
                            for t in threats[:3]:
                                print(f"  • {t.url[:60]}... (Risk: {t.risk_score}/100)")
                            if len(threats) > 3:
                                print(f"  ... and {len(threats) - 3} more")
                    else:
                        avg_risk = np.mean([t.risk_score for t in threats])
                        color = Colors.FAIL if avg_risk > 70 else Colors.WARNING if avg_risk > 45 else Colors.OKGREEN
                        
                        print(f"\n{color}{campaign_id}{Colors.ENDC} ({len(threats)} threats)")
                        print(f"  Average Risk: {avg_risk:.1f}/100")
                        print(f"  Threats:")
                        for t in threats[:5]:
                            print(f"    • {t.url[:60]}... (Risk: {t.risk_score}/100)")
                        if len(threats) > 5:
                            print(f"    ... and {len(threats) - 5} more")
                
                # Save campaign report
                report = campaign_detector.generate_campaign_report(campaigns)
                report_file = f"campaign_report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
                report_path = os.path.join(Config.CAMPAIGN_DATA, report_file)
                with open(report_path, 'w', encoding='utf-8') as f:
                    f.write(report)
                Colors.print_success(f"\n✓ Campaign report saved: {report_file}")
                
                # Visualize campaigns
                valid_dnas = [d for d in collected_dnas if d.combined_embedding is not None]
                if len(valid_dnas) >= 3:
                    try:
                        Colors.print_info("\n📊 Generating campaign visualization...")
                        
                        embeddings = np.array([d.combined_embedding for d in valid_dnas])
                        
                        # t-SNE for 2D visualization
                        tsne = TSNE(
                            n_components=2,
                            random_state=Config.RANDOM_STATE,
                            perplexity=min(30, len(embeddings) - 1)
                        )
                        embeddings_2d = tsne.fit_transform(embeddings)
                        
                        plt.figure(figsize=(14, 10))
                        
                        for campaign_id, threats in campaigns.items():
                            indices = [valid_dnas.index(t) for t in threats if t in valid_dnas]
                            if not indices:
                                continue
                            
                            points = embeddings_2d[indices]
                            
                            if campaign_id == 'UNCLUSTERED':
                                plt.scatter(points[:, 0], points[:, 1], c='gray',
                                          label=campaign_id, alpha=0.3, s=50)
                            else:
                                color = plt.cm.Set1(hash(campaign_id) % 9)
                                plt.scatter(points[:, 0], points[:, 1], c=[color],
                                          label=campaign_id, alpha=0.7, s=100, edgecolors='black')
                        
                        plt.title('Campaign Clustering Visualization (t-SNE)',
                                fontsize=14, fontweight='bold')
                        plt.xlabel('Dimension 1')
                        plt.ylabel('Dimension 2')
                        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
                        plt.grid(True, alpha=0.3)
                        plt.tight_layout()
                        
                        viz_file = f'campaign_viz_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png'
                        viz_path = os.path.join(Config.CAMPAIGN_DATA, viz_file)
                        plt.savefig(viz_path, dpi=Config.VIZ_DPI, bbox_inches='tight')
                        plt.close()
                        
                        Colors.print_success(f"✓ Visualization saved: {viz_file}")
                        Colors.print_info("✓ Opening visualization in browser...")
                        try:
                            webbrowser.open(f'file://{os.path.abspath(viz_path)}')
                            Colors.print_success("✓ Visualization opened successfully")
                        except Exception as e:
                            Colors.print_warning(f"Could not auto-open: {e}")
                            Colors.print_info(f"Manual path: {viz_path}")
                    except Exception as e:
                        Colors.print_warning(f"Visualization failed: {e}")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            elif choice == '17':
                Colors.print_header("📊 THREAT DNA DATABASE STATS")
                
                print(f"\n{Colors.BOLD}━━━ Collection Statistics ━━━{Colors.ENDC}")
                print(f"  DNA Samples Collected:  {len(collected_dnas)}")
                
                if vector_db:
                    print(f"  Vector DB Entries:      {vector_db.get_threat_count()}")
                else:
                    print(f"  Vector DB:              Not available")
                
                if collected_dnas:
                    with_embeddings = sum(1 for d in collected_dnas if d.combined_embedding is not None)
                    with_screenshots = sum(1 for d in collected_dnas if d.screenshot_path)
                    with_dom = sum(1 for d in collected_dnas if d.dom_snapshot)
                    
                    print(f"  With Embeddings:        {with_embeddings} ({with_embeddings/len(collected_dnas)*100:.1f}%)")
                    print(f"  With Screenshots:       {with_screenshots} ({with_screenshots/len(collected_dnas)*100:.1f}%)")
                    print(f"  With DOM Snapshots:     {with_dom} ({with_dom/len(collected_dnas)*100:.1f}%)")
                    
                    avg_risk = np.mean([d.risk_score for d in collected_dnas])
                    print(f"  Average Risk Score:     {avg_risk:.1f}/100")
                    
                    print(f"\n{Colors.BOLD}━━━ Risk Distribution ━━━{Colors.ENDC}")
                    high = sum(1 for d in collected_dnas if d.risk_score >= Config.HIGH_RISK)
                    medium = sum(1 for d in collected_dnas if Config.MEDIUM_RISK <= d.risk_score < Config.HIGH_RISK)
                    low = len(collected_dnas) - high - medium
                    
                    print(f"  {Colors.FAIL}High Risk:    {high:3d} ({high/len(collected_dnas)*100:5.1f}%){Colors.ENDC}")
                    print(f"  {Colors.WARNING}Medium Risk:  {medium:3d} ({medium/len(collected_dnas)*100:5.1f}%){Colors.ENDC}")
                    print(f"  {Colors.OKGREEN}Low Risk:     {low:3d} ({low/len(collected_dnas)*100:5.1f}%){Colors.ENDC}")
                    
                    print(f"\n{Colors.BOLD}━━━ Recent DNA Samples ━━━{Colors.ENDC}")
                    for dna in collected_dnas[-5:]:
                        print(f"  • {dna.threat_id}: {dna.url[:50]}... (Risk: {dna.risk_score}/100)")
                
                else:
                    print(f"\n{Colors.WARNING}No DNA samples collected yet.{Colors.ENDC}")
                    print(f"Use option [14] to start collecting threat DNA!")
                
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
            
            else:
                Colors.print_warning("Invalid option. Please try again.")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")
    
    except KeyboardInterrupt:
        print(f"\n\n{Colors.WARNING}Operation cancelled by user{Colors.ENDC}")
        monitor.stop_monitoring()
    except Exception as e:
        Colors.print_error(f"Fatal error: {e}")
        logger.error(f"Fatal error: {e}", exc_info=True)
        import traceback
        traceback.print_exc()
    finally:
        if monitor.is_monitoring:
            monitor.stop_monitoring()
            Colors.print_info("Monitoring stopped")


if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print(f"\n{Colors.FAIL}Critical error: {e}{Colors.ENDC}")
        import traceback
        traceback.print_exc()
        input("\nPress Enter to exit...")
```
