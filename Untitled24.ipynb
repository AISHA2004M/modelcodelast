```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
PHISHING DETECTION SYSTEM v10.0 - FIXED EDITION
==================================================
CHANGES IN THIS VERSION:
  1. Suspicious/phishing URLs are BLOCKED from opening in browser
     (analyzed fully, but browser.open is suppressed for risk >= MEDIUM_RISK)
  2. Menu renumbered sequentially with no gaps:
       [1]  Quick URL Scan
       [2]  Full Analysis with Report
       [3]  Threat Intel Stats
       [4]  Reload Threat Feeds
       [5]  Help & Information
       [6]  3D Cyber Map
       [7]  Real-Time Dashboard
       [8]  AI Chat Assistant
       [9]  Analytics & Trends
       [10] Deep Crawl & DNA Collection
       [11] Find Similar Threats
       [12] Campaign Detection
       [13] DNA Database Stats
       [0]  Exit
  3. Risk score clamped to 0-100 integer (no float overflow)
  4. Stale model auto-detection: if saved model was built with a different
     sklearn version it is deleted and retrained automatically.

INSTALLATION:
pip install playwright sentence-transformers chromadb scikit-learn pandas numpy \
    matplotlib seaborn psutil requests tqdm pillow opencv-python networkx joblib
playwright install chromium
"""

import os
import re
import json
import logging
import hashlib
import warnings
import webbrowser
import requests
import threading
import time
import psutil
import asyncio
from datetime import datetime
from urllib.parse import urlparse
from collections import Counter, deque
from dataclasses import dataclass, asdict, field
from typing import Dict, List, Optional, Tuple, Any
from pathlib import Path

import numpy as np
import pandas as pd
from tqdm import tqdm

import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, IsolationForest
from sklearn.metrics import accuracy_score
from sklearn.cluster import DBSCAN
from sklearn.manifold import TSNE

try:
    import cv2
    HAS_CV = True
except:
    HAS_CV = False

try:
    import networkx as nx
    HAS_NETWORKX = True
except:
    HAS_NETWORKX = False

try:
    from playwright.async_api import async_playwright
    HAS_PLAYWRIGHT = True
except:
    HAS_PLAYWRIGHT = False

try:
    from sentence_transformers import SentenceTransformer
    HAS_TRANSFORMERS = True
except:
    HAS_TRANSFORMERS = False

try:
    import chromadb
    HAS_VECTORDB = True
except:
    HAS_VECTORDB = False

warnings.filterwarnings('ignore')
sns.set_style("whitegrid")


# ==================== COLORS ====================

class Colors:
    HEADER  = '\033[95m'
    OKBLUE  = '\033[94m'
    OKCYAN  = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL    = '\033[91m'
    ENDC    = '\033[0m'
    BOLD    = '\033[1m'

    @staticmethod
    def print_header(text):
        print(f"\n{Colors.BOLD}{Colors.HEADER}{'='*80}\n  {text}\n{'='*80}{Colors.ENDC}\n")

    @staticmethod
    def print_success(text):
        print(f"{Colors.OKGREEN}✓ {text}{Colors.ENDC}")

    @staticmethod
    def print_error(text):
        print(f"{Colors.FAIL}✗ {text}{Colors.ENDC}")

    @staticmethod
    def print_info(text):
        print(f"{Colors.OKCYAN}ℹ {text}{Colors.ENDC}")

    @staticmethod
    def print_warning(text):
        print(f"{Colors.WARNING}⚠ {text}{Colors.ENDC}")


# ==================== CONFIGURATION ====================

class Config:
    OUT_DIR = os.path.join(os.path.expanduser("~"), "Desktop", "phishing_detection_v10")

    POSSIBLE_DATA_PATHS = [
        os.path.join(os.path.expanduser("~"), "Desktop", "datasets", "final_cleaned_preprocessed.csv"),
        os.path.join(os.path.expanduser("~"), "Desktop", "phishing_dataset.csv"),
        os.path.join(os.path.expanduser("~"), "Downloads", "phishing_dataset.csv"),
    ]
    DATA_FILE  = None
    MODEL_FILE = os.path.join(OUT_DIR, "phishing_model.pkl")

    PLOTS_DIR         = os.path.join(OUT_DIR, "analysis_plots")
    REPORTS_DIR       = os.path.join(OUT_DIR, "reports")
    THREAT_INTEL_CACHE= os.path.join(OUT_DIR, "threat_intelligence")
    SCREENSHOTS_DIR   = os.path.join(OUT_DIR, "screenshots")
    VISUALIZATION_DIR = os.path.join(OUT_DIR, "visualizations")
    CYBER_MAP_DIR     = os.path.join(OUT_DIR, "cyber_maps")
    DASHBOARD_DIR     = os.path.join(OUT_DIR, "dashboards")
    THREAT_DNA_DIR    = os.path.join(OUT_DIR, "threat_dna")
    CRAWL_ARTIFACTS   = os.path.join(THREAT_DNA_DIR, "artifacts")
    HAR_LOGS          = os.path.join(THREAT_DNA_DIR, "har_logs")
    HTML_DUMPS        = os.path.join(THREAT_DNA_DIR, "html_dumps")
    VECTOR_DB_DIR     = os.path.join(THREAT_DNA_DIR, "vector_db")
    CAMPAIGN_DATA     = os.path.join(THREAT_DNA_DIR, "campaigns")
    LOG_FILE          = os.path.join(OUT_DIR, "phishing.log")

    TEST_SIZE    = 0.25
    RANDOM_STATE = 42
    HIGH_RISK    = 70
    MEDIUM_RISK  = 45

    WEIGHTS = {
        'rf': 0.20, 'gb': 0.20, 'adversarial_rf': 0.15,
        'threat_intel': 0.15, 'evasion_detector': 0.10, 'threat_dna': 0.20
    }

    ADVERSARIAL_ENABLED        = True
    ADVERSARIAL_NOISE_EPSILON  = 0.1
    ADVERSARIAL_TRAINING_RATIO = 0.3
    EVASION_THRESHOLD          = 0.7
    MONITORING_INTERVAL        = 5
    MAX_HISTORY                = 100
    CHAT_HISTORY_SIZE          = 50
    CRAWL_TIMEOUT              = 45000
    CRAWL_WAIT_TIME            = 3000
    ENABLE_SCREENSHOTS         = True
    ENABLE_HAR_CAPTURE         = False
    ENABLE_DOM_SNAPSHOT        = True
    TEXT_EMBEDDING_MODEL       = "all-MiniLM-L6-v2"
    EMBEDDING_DIM              = 384
    SIMILARITY_THRESHOLD       = 0.85
    CAMPAIGN_CLUSTERING_EPS    = 0.3
    CAMPAIGN_MIN_SAMPLES       = 3
    VIZ_DPI                    = 150

    @classmethod
    def validate(cls):
        for d in [cls.OUT_DIR, cls.PLOTS_DIR, cls.REPORTS_DIR, cls.THREAT_INTEL_CACHE,
                  cls.SCREENSHOTS_DIR, cls.VISUALIZATION_DIR, cls.CYBER_MAP_DIR,
                  cls.DASHBOARD_DIR, cls.THREAT_DNA_DIR, cls.CRAWL_ARTIFACTS,
                  cls.HAR_LOGS, cls.HTML_DUMPS, cls.VECTOR_DB_DIR, cls.CAMPAIGN_DATA]:
            os.makedirs(d, exist_ok=True)
        for path in cls.POSSIBLE_DATA_PATHS:
            if os.path.exists(path):
                cls.DATA_FILE = path
                break
        return cls.DATA_FILE is not None


Config.validate()

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[logging.FileHandler(Config.LOG_FILE), logging.StreamHandler()]
)
logger = logging.getLogger(__name__)


# ==================== THREAT DNA ====================

@dataclass
class ThreatDNA:
    url: str
    timestamp: datetime
    threat_id: str = field(default="")
    screenshot_path: Optional[str] = None
    html_content: Optional[str] = None
    console_logs: List[str] = field(default_factory=list)
    dom_snapshot: Optional[Dict] = None
    network_requests: List[Dict] = field(default_factory=list)
    text_content: Optional[str] = None
    extracted_links: List[str] = field(default_factory=list)
    extracted_forms: List[Dict] = field(default_factory=list)
    javascript_behavior: Dict = field(default_factory=dict)
    text_embedding: Optional[np.ndarray] = None
    image_embedding: Optional[np.ndarray] = None
    dom_embedding: Optional[np.ndarray] = None
    combined_embedding: Optional[np.ndarray] = None
    page_title: Optional[str] = None
    response_time_ms: Optional[float] = None
    risk_score: float = 0.0
    threat_category: Optional[str] = None
    campaign_id: Optional[str] = None
    similar_threats: List[str] = field(default_factory=list)

    def __post_init__(self):
        if not self.threat_id:
            self.threat_id = self.generate_threat_id()

    def generate_threat_id(self) -> str:
        components = f"{self.url}{self.timestamp.isoformat()}".encode('utf-8')
        return f"THR-{hashlib.sha256(components).hexdigest()[:16].upper()}"

    def get_fingerprint(self) -> str:
        components = [self.url, self.text_content or "",
                      str(self.dom_snapshot) if self.dom_snapshot else "",
                      str(self.javascript_behavior)]
        return hashlib.sha256("".join(components).encode('utf-8')).hexdigest()

    def to_dict(self) -> Dict:
        data = asdict(self)
        for key in ['text_embedding', 'image_embedding', 'dom_embedding', 'combined_embedding']:
            if data[key] is not None and isinstance(data[key], np.ndarray):
                data[key] = data[key].tolist()
        data['timestamp'] = self.timestamp.isoformat()
        return data


# ==================== CRAWLER ====================

class ThreatDNACrawler:
    def __init__(self):
        self.storage_dir    = Path(Config.CRAWL_ARTIFACTS)
        self.screenshots_dir= Path(Config.SCREENSHOTS_DIR)
        self.html_dir       = Path(Config.HTML_DUMPS)
        logger.info("✓ Threat DNA Crawler initialized")

    async def crawl_url(self, url: str) -> Optional[ThreatDNA]:
        if not HAS_PLAYWRIGHT:
            Colors.print_error("Playwright not installed!")
            return None

        Colors.print_info(f"Crawling: {url[:60]}...")
        start_time = datetime.now()
        dna = ThreatDNA(url=url, timestamp=start_time)
        browser = None
        try:
            async with async_playwright() as p:
                browser = await p.chromium.launch(
                    headless=True,
                    args=['--no-sandbox','--disable-setuid-sandbox',
                          '--disable-dev-shm-usage','--disable-gpu']
                )
                context = await browser.new_context(
                    viewport={'width': 1920, 'height': 1080},
                    user_agent='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
                    ignore_https_errors=True
                )
                context.set_default_timeout(Config.CRAWL_TIMEOUT)
                page = await context.new_page()
                page.on("console", lambda msg: dna.console_logs.append(f"[{msg.type}] {msg.text}"))
                page.on("request", lambda req: dna.network_requests.append({
                    'url': req.url, 'method': req.method, 'timestamp': datetime.now().isoformat()
                }))
                try:
                    response = await page.goto(url, timeout=Config.CRAWL_TIMEOUT, wait_until='domcontentloaded')
                    if response is None:
                        raise Exception("No response received")
                    dna.response_time_ms = (datetime.now() - start_time).total_seconds() * 1000
                    try:
                        await page.wait_for_load_state('networkidle', timeout=Config.CRAWL_WAIT_TIME)
                    except:
                        pass
                    dna.page_title   = await page.title()
                    dna.html_content = await page.content()
                    if Config.ENABLE_SCREENSHOTS:
                        try:
                            sp = self.screenshots_dir / f"{dna.threat_id}.png"
                            await page.screenshot(path=str(sp), full_page=False, timeout=10000)
                            dna.screenshot_path = str(sp)
                        except Exception as e:
                            logger.warning(f"Screenshot failed: {e}")
                    if Config.ENABLE_DOM_SNAPSHOT:
                        try:
                            dna.dom_snapshot = await page.evaluate('''() => {
                                try {
                                    return {
                                        title: document.title||'',
                                        forms: Array.from(document.forms||[]).slice(0,10).map(f=>({
                                            action:f.action||'',method:f.method||'',
                                            inputs:Array.from(f.elements||[]).slice(0,20).map(e=>({type:e.type||'',name:e.name||''}))
                                        })),
                                        links:Array.from(document.links||[]).slice(0,50).map(l=>l.href||''),
                                        scripts:Array.from(document.scripts||[]).slice(0,20).map(s=>s.src||''),
                                        iframes:Array.from(document.querySelectorAll('iframe')||[]).slice(0,10).map(i=>i.src||'')
                                    }
                                } catch(e){return{title:'',forms:[],links:[],scripts:[],iframes:[]};}
                            }''')
                        except Exception as e:
                            logger.warning(f"DOM snapshot failed: {e}")
                            dna.dom_snapshot = {'forms':[],'links':[],'scripts':[],'iframes':[]}
                    try:
                        tc = await page.evaluate('()=>document.body?document.body.innerText:""')
                        dna.text_content = tc[:10000] if tc else ""
                    except:
                        dna.text_content = ""
                    if dna.dom_snapshot:
                        dna.extracted_links = dna.dom_snapshot.get('links', [])
                        dna.extracted_forms = dna.dom_snapshot.get('forms', [])
                    try:
                        dna.javascript_behavior = await page.evaluate('''()=>{
                            try{return{hasOnLoad:!!window.onload,hasLocalStorage:typeof(Storage)!=='undefined',
                                hasGeolocation:'geolocation'in navigator,windowProperties:Object.keys(window).length,
                                hasAlert:typeof alert!=='undefined',hasPrompt:typeof prompt!=='undefined'
                            }}catch(e){return{};}
                        }''')
                    except:
                        dna.javascript_behavior = {}
                    Colors.print_success("✓ Crawl completed")
                except Exception as e:
                    err = str(e)
                    if "ERR_NAME_NOT_RESOLVED" in err:
                        Colors.print_error("DNS resolution failed - domain doesn't exist")
                    elif "ERR_CONNECTION_REFUSED" in err:
                        Colors.print_error("Connection refused")
                    elif "Timeout" in err:
                        Colors.print_error("Connection timeout")
                    elif "SSL" in err:
                        Colors.print_error("SSL/TLS error")
                    else:
                        Colors.print_error(f"Navigation failed: {err[:80]}")
                    return None
                if dna.html_content:
                    try:
                        hp = self.html_dir / f"{dna.threat_id}.html"
                        with open(hp,'w',encoding='utf-8',errors='ignore') as f:
                            f.write(dna.html_content)
                    except:
                        pass
                await context.close()
                await browser.close()
            return dna
        except Exception as e:
            logger.error(f"Fatal crawl error: {e}")
            if browser:
                try: await browser.close()
                except: pass
            return None

    def crawl_url_sync(self, url: str) -> Optional[ThreatDNA]:
        try:
            try:
                old = asyncio.get_event_loop()
                if old.is_running():
                    import concurrent.futures
                    with concurrent.futures.ThreadPoolExecutor() as ex:
                        return ex.submit(self._run_in_new_loop, url).result(timeout=60)
            except RuntimeError:
                pass
            loop = asyncio.new_event_loop()
            asyncio.set_event_loop(loop)
            try:
                return loop.run_until_complete(self.crawl_url(url))
            finally:
                try:
                    pending = asyncio.all_tasks(loop)
                    for t in pending: t.cancel()
                    loop.run_until_complete(asyncio.gather(*pending, return_exceptions=True))
                except: pass
                try: loop.close()
                except: pass
        except Exception as e:
            logger.error(f"Sync crawl error: {e}")
            Colors.print_error(f"Crawl error: {str(e)[:100]}")
            return None

    def _run_in_new_loop(self, url: str) -> Optional[ThreatDNA]:
        loop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)
        try:
            return loop.run_until_complete(self.crawl_url(url))
        finally:
            loop.close()


# ==================== EMBEDDINGS ====================

class EmbeddingGenerator:
    def __init__(self):
        self.text_model = None
        if HAS_TRANSFORMERS:
            try:
                Colors.print_info("Loading text embedding model...")
                self.text_model = SentenceTransformer(Config.TEXT_EMBEDDING_MODEL)
                Colors.print_success("✓ Text embedding model loaded")
            except Exception as e:
                Colors.print_warning(f"Text model load failed: {e}")

    def generate_text_embedding(self, text: str) -> Optional[np.ndarray]:
        if not self.text_model or not text or len(text.strip()) < 10:
            return None
        try:
            return np.array(self.text_model.encode(text[:5000], show_progress_bar=False, convert_to_numpy=True))
        except Exception as e:
            logger.warning(f"Text embedding failed: {e}")
            return None

    def generate_image_embedding(self, image_path: str) -> Optional[np.ndarray]:
        if not HAS_CV or not image_path or not os.path.exists(image_path):
            return None
        try:
            img = cv2.imread(image_path)
            if img is None or img.size == 0: return None
            h,w = img.shape[:2]
            if max(h,w) > 1024:
                sc = 1024/max(h,w)
                img = cv2.resize(img,(int(w*sc),int(h*sc)))
            hist = np.concatenate([
                cv2.calcHist([img],[i],None,[256],[0,256]).flatten() for i in range(3)
            ])
            hist = hist/(hist.sum()+1e-7)
            if len(hist)>Config.EMBEDDING_DIM:
                hist=hist[::len(hist)//Config.EMBEDDING_DIM][:Config.EMBEDDING_DIM]
            if len(hist)<Config.EMBEDDING_DIM:
                hist=np.pad(hist,(0,Config.EMBEDDING_DIM-len(hist)))
            return hist
        except Exception as e:
            logger.warning(f"Image embedding failed: {e}")
            return None

    def generate_dom_embedding(self, dom_snapshot: Dict) -> Optional[np.ndarray]:
        if not dom_snapshot: return None
        try:
            features = [
                len(dom_snapshot.get('forms',[])),
                len(dom_snapshot.get('links',[])),
                len(dom_snapshot.get('scripts',[])),
                len(dom_snapshot.get('iframes',[])),
                sum(1 for f in dom_snapshot.get('forms',[]) if f.get('method','').lower()=='post'),
                sum(1 for l in dom_snapshot.get('links',[]) if str(l).startswith('javascript:')),
                sum(1 for s in dom_snapshot.get('scripts',[]) if s),
                sum(1 for i in dom_snapshot.get('iframes',[]) if i)
            ]
            emb = np.array(features+[0]*(Config.EMBEDDING_DIM-len(features)))
            return emb[:Config.EMBEDDING_DIM]
        except Exception as e:
            logger.warning(f"DOM embedding failed: {e}")
            return None

    def generate_combined_embedding(self, dna: ThreatDNA) -> Optional[np.ndarray]:
        embeddings,weights=[],[]
        if dna.text_embedding  is not None: embeddings.append(dna.text_embedding);  weights.append(0.5)
        if dna.image_embedding is not None: embeddings.append(dna.image_embedding); weights.append(0.3)
        if dna.dom_embedding   is not None: embeddings.append(dna.dom_embedding);   weights.append(0.2)
        if not embeddings: return None
        try:
            w = np.array(weights)/sum(weights)
            combined = np.average(embeddings, axis=0, weights=w)
            n = np.linalg.norm(combined)
            return combined/n if n>0 else combined
        except Exception as e:
            logger.warning(f"Combined embedding failed: {e}")
            return None

    def enrich_threat_dna(self, dna: ThreatDNA) -> ThreatDNA:
        try:
            if dna.text_content and len(dna.text_content.strip())>10:
                dna.text_embedding  = self.generate_text_embedding(dna.text_content)
            if dna.screenshot_path and os.path.exists(dna.screenshot_path):
                dna.image_embedding = self.generate_image_embedding(dna.screenshot_path)
            if dna.dom_snapshot:
                dna.dom_embedding   = self.generate_dom_embedding(dna.dom_snapshot)
            dna.combined_embedding  = self.generate_combined_embedding(dna)
        except Exception as e:
            logger.warning(f"DNA enrichment error: {e}")
        return dna


# ==================== VECTOR DB ====================

class ThreatVectorDB:
    def __init__(self):
        self.client=None; self.collection=None
        if HAS_VECTORDB:
            try:
                self.client     = chromadb.PersistentClient(path=Config.VECTOR_DB_DIR)
                self.collection = self.client.get_or_create_collection(name="threat_dna")
                Colors.print_success(f"✓ Vector DB initialized ({self.collection.count()} entries)")
            except Exception as e:
                Colors.print_warning(f"Vector DB init failed: {e}")

    def add_threat(self, dna: ThreatDNA) -> bool:
        if not self.collection or dna.combined_embedding is None: return False
        try:
            try:
                ex = self.collection.get(ids=[dna.threat_id])
                if ex['ids']: self.collection.delete(ids=[dna.threat_id])
            except: pass
            self.collection.add(
                ids=[dna.threat_id],
                embeddings=[dna.combined_embedding.tolist()],
                metadatas=[{'url':dna.url[:500],'timestamp':dna.timestamp.isoformat(),
                            'risk_score':float(dna.risk_score),'threat_category':dna.threat_category or 'unknown'}]
            )
            return True
        except Exception as e:
            logger.error(f"Vector DB add failed: {e}"); return False

    def find_similar_threats(self, dna: ThreatDNA, top_k:int=5) -> List[Dict]:
        if not self.collection or dna.combined_embedding is None: return []
        try:
            results = self.collection.query(
                query_embeddings=[dna.combined_embedding.tolist()],
                n_results=min(top_k+1, self.collection.count())
            )
            out=[]
            if results['ids']:
                for i,tid in enumerate(results['ids'][0]):
                    if tid!=dna.threat_id:
                        sim=1-results['distances'][0][i]
                        if sim>0.5:
                            out.append({'threat_id':tid,'similarity':sim,'metadata':results['metadatas'][0][i]})
            return out[:top_k]
        except Exception as e:
            logger.error(f"Vector search failed: {e}"); return []

    def get_threat_count(self) -> int:
        if not self.collection: return 0
        try: return self.collection.count()
        except: return 0


# ==================== CAMPAIGN DETECTOR ====================

class CampaignDetector:
    def __init__(self):
        self.campaigns={}; self.next_campaign_id=1

    def detect_campaigns(self, threat_dnas: List[ThreatDNA]) -> Dict[str,List[ThreatDNA]]:
        if len(threat_dnas)<Config.CAMPAIGN_MIN_SAMPLES:
            return {'UNCLUSTERED':threat_dnas}
        embeddings,valid=[],[]
        for dna in threat_dnas:
            if dna.combined_embedding is not None and len(dna.combined_embedding)==Config.EMBEDDING_DIM:
                embeddings.append(dna.combined_embedding); valid.append(dna)
        if len(embeddings)<Config.CAMPAIGN_MIN_SAMPLES:
            return {'UNCLUSTERED':threat_dnas}
        try:
            X=np.array(embeddings)
            X=X/(np.linalg.norm(X,axis=1,keepdims=True)+1e-8)
            eps=Config.CAMPAIGN_CLUSTERING_EPS if len(embeddings)>=10 else 0.4
            labels=DBSCAN(eps=eps,min_samples=min(Config.CAMPAIGN_MIN_SAMPLES,len(embeddings)//3),metric='cosine').fit(X).labels_
            campaigns={}
            for idx,label in enumerate(labels):
                cid='UNCLUSTERED' if label==-1 else f'CAMPAIGN-{label+1}'
                campaigns.setdefault(cid,[]).append(valid[idx])
                valid[idx].campaign_id=cid
            extras=[d for d in threat_dnas if d not in valid]
            if extras: campaigns.setdefault('UNCLUSTERED',[]).extend(extras)
            return campaigns
        except Exception as e:
            logger.error(f"Campaign detection failed: {e}")
            return {'UNCLUSTERED':threat_dnas}

    def generate_campaign_report(self, campaigns: Dict[str,List[ThreatDNA]]) -> str:
        r =f"\n{'='*80}\nCAMPAIGN DETECTION REPORT\n{'='*80}\n\n"
        r+=f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n"
        r+=f"Campaigns: {len([c for c in campaigns if c!='UNCLUSTERED'])}\n"
        r+=f"Unclustered: {len(campaigns.get('UNCLUSTERED',[]))}\n\n"
        for cid,threats in sorted([(k,v) for k,v in campaigns.items() if k!='UNCLUSTERED'],key=lambda x:len(x[1]),reverse=True):
            r+=f"\n{cid} ({len(threats)} threats)\n{'='*60}\n"
            if threats:
                r+=f"Avg Risk: {np.mean([t.risk_score for t in threats]):.1f}/100\n"
                for t in threats[:10]: r+=f"  - {t.url[:70]}... (Risk: {t.risk_score}/100)\n"
                if len(threats)>10: r+=f"  ... and {len(threats)-10} more\n"
        if campaigns.get('UNCLUSTERED'):
            r+=f"\nUNCLUSTERED ({len(campaigns['UNCLUSTERED'])})\n{'='*60}\n"
            for t in campaigns['UNCLUSTERED'][:5]: r+=f"  - {t.url[:70]}... (Risk: {t.risk_score}/100)\n"
        return r


# ==================== SUPPORTING CLASSES ====================

class SecurityChatAssistant:
    def __init__(self):
        self.chat_history=deque(maxlen=Config.CHAT_HISTORY_SIZE)
        self.context={'recent_detections':[],'threat_stats':{},'system_health':{}}

    def process_query(self, query:str)->str:
        q=query.lower().strip()
        self.chat_history.append({'timestamp':datetime.now(),'query':query,'type':'user'})
        r=self._generate_response(q)
        self.chat_history.append({'timestamp':datetime.now(),'response':r,'type':'assistant'})
        return r

    def _generate_response(self, q:str)->str:
        if 'statistics' in q or 'stats' in q:
            tot=len(self.context.get('recent_detections',[]))
            phi=sum(1 for d in self.context.get('recent_detections',[]) if d.get('is_phishing'))
            return f"📊 {tot} total scans, {phi} phishing ({phi/tot*100 if tot else 0:.1f}%)"
        if 'help'    in q: return "🤖 Ask me about: statistics, threats, protection tips, system status."
        if 'threat'  in q or 'phishing' in q: return "🎣 Phishing uses disguised communication to steal sensitive info. Always verify before clicking!"
        if 'status'  in q or 'health'   in q:
            h=self.context.get('system_health',{})
            return f"⚙️ CPU: {h.get('cpu',0):.1f}% | Memory: {h.get('memory',0):.1f}%"
        if 'protect' in q or 'safe'     in q: return "🛡️ Tips: 1) Verify sender  2) Hover links  3) Check HTTPS  4) Enable 2FA"
        if 'latest'  in q or 'recent'   in q:
            rec=self.context.get('recent_detections',[])[-5:]
            return f"📋 Last {len(rec)} scans: {sum(1 for d in rec if d.get('is_phishing'))} threats" if rec else "No recent scans"
        return "I'm your AI security assistant! Ask me about statistics, threats, protection tips, or status."

    def update_context(self, detections=None, threat_stats=None, system_health=None):
        if detections:    self.context['recent_detections']=detections
        if threat_stats:  self.context['threat_stats']=threat_stats
        if system_health: self.context['system_health']=system_health


class RealTimeMonitor:
    def __init__(self):
        self.detection_history=deque(maxlen=Config.MAX_HISTORY)
        self.threat_timeline=deque(maxlen=Config.MAX_HISTORY)
        self.system_metrics=deque(maxlen=Config.MAX_HISTORY)
        self.is_monitoring=False; self.monitor_thread=None

    def start_monitoring(self):
        if self.is_monitoring: return
        self.is_monitoring=True
        self.monitor_thread=threading.Thread(target=self._monitoring_loop,daemon=True)
        self.monitor_thread.start()
        Colors.print_success("✓ Real-time monitoring started")

    def stop_monitoring(self):
        self.is_monitoring=False
        if self.monitor_thread: self.monitor_thread.join(timeout=2)

    def _monitoring_loop(self):
        while self.is_monitoring:
            try:
                self.system_metrics.append({
                    'timestamp':datetime.now(),
                    'cpu_percent':psutil.cpu_percent(interval=1),
                    'memory_percent':psutil.virtual_memory().percent,
                    'active_threats':len([d for d in self.detection_history if d.get('is_phishing')])
                })
                time.sleep(Config.MONITORING_INTERVAL)
            except Exception as e:
                logger.error(f"Monitoring error: {e}"); time.sleep(Config.MONITORING_INTERVAL)

    def log_detection(self, r:Dict):
        ev={'timestamp':datetime.now(),'url':r.get('url','')[:50],
            'is_phishing':r.get('is_phishing',False),'risk_score':r.get('risk_score',0),
            'risk_level':r.get('risk_level','UNKNOWN')}
        self.detection_history.append(ev)
        if ev['is_phishing']: self.threat_timeline.append(ev)

    def get_current_stats(self)->Dict:
        tot=len(self.detection_history)
        phi=sum(1 for d in self.detection_history if d['is_phishing'])
        lm=self.system_metrics[-1] if self.system_metrics else {}
        return {'total_scans':tot,'phishing_detected':phi,'safe_urls':tot-phi,
                'detection_rate':(phi/tot*100) if tot>0 else 0,
                'cpu_usage':lm.get('cpu_percent',0),'memory_usage':lm.get('memory_percent',0),
                'active_threats':lm.get('active_threats',0)}


class CyberMapVisualizer:
    def __init__(self):
        self.country_codes={
            'US':(37.09,-95.71),'CN':(35.86,104.19),'RU':(61.52,105.31),
            'BR':(-14.23,-51.92),'IN':(20.59,78.96),'UK':(55.37,-3.43),
            'DE':(51.16,10.45),'FR':(46.22,2.21),'CA':(56.13,-106.34),
            'AU':(-25.27,133.77),'JP':(36.20,138.25),'Unknown':(0,0)
        }

    def create_3d_cyber_map(self, threat_data:List[Dict])->Optional[str]:
        if not threat_data: return None
        try:
            fig=plt.figure(figsize=(16,12))
            ax1=fig.add_subplot(221,projection='3d')
            u=np.linspace(0,2*np.pi,50); v=np.linspace(0,np.pi,50)
            ax1.plot_surface(np.outer(np.cos(u),np.sin(v)),np.outer(np.sin(u),np.sin(v)),
                             np.outer(np.ones(50),np.cos(v)),color='lightblue',alpha=0.3,linewidth=0)
            for t in threat_data[-50:]:
                lat,lon=self._get_threat_location(t)
                lr,lo=np.radians(lat),np.radians(lon)
                risk=t.get('risk_score',50)/100
                ax1.scatter([np.cos(lr)*np.cos(lo)],[np.cos(lr)*np.sin(lo)],[np.sin(lr)],
                            c='red' if risk>0.7 else 'orange' if risk>0.45 else 'yellow',
                            s=50+risk*100,alpha=0.8,edgecolors='black')
            ax1.set_title('3D Global Threat Map',fontweight='bold',fontsize=12)
            ax1.set_box_aspect([1,1,1])

            ax2=fig.add_subplot(222)
            cc=Counter([self._guess_country(t) for t in threat_data])
            top=dict(sorted(cc.items(),key=lambda x:x[1],reverse=True)[:10])
            ax2.barh(list(top.keys()),list(top.values()),color=plt.cm.YlOrRd(np.linspace(0.3,1,len(top))))
            ax2.set_xlabel('Threat Count'); ax2.set_title('Top 10 Threat Origins',fontweight='bold'); ax2.grid(axis='x',alpha=0.3)

            ax3=fig.add_subplot(223)
            rs=[t.get('risk_score',0) for t in threat_data]
            ax3.hist(rs,bins=20,color='coral',edgecolor='black',alpha=0.7)
            ax3.axvline(x=Config.HIGH_RISK,color='red',linestyle='--',label='High Risk')
            ax3.axvline(x=Config.MEDIUM_RISK,color='orange',linestyle='--',label='Medium Risk')
            ax3.set_xlabel('Risk Score'); ax3.set_ylabel('Frequency')
            ax3.set_title('Risk Score Distribution',fontweight='bold'); ax3.legend(); ax3.grid(alpha=0.3)

            ax4=fig.add_subplot(224)
            if len(threat_data)>1:
                cols=['red' if r>Config.HIGH_RISK else 'orange' if r>Config.MEDIUM_RISK else 'green' for r in rs]
                ax4.scatter(range(len(rs)),rs,c=cols,alpha=0.6,s=50)
                ax4.plot(range(len(rs)),rs,'gray',alpha=0.3,linewidth=1)
                ax4.axhline(y=Config.HIGH_RISK,color='red',linestyle='--',alpha=0.5)
                ax4.axhline(y=Config.MEDIUM_RISK,color='orange',linestyle='--',alpha=0.5)
                ax4.set_xlabel('Detection Sequence'); ax4.set_ylabel('Risk Score')
                ax4.set_title('Threat Timeline',fontweight='bold'); ax4.grid(alpha=0.3)

            plt.suptitle('CYBER THREAT MAP',fontsize=16,fontweight='bold'); plt.tight_layout()
            fp=os.path.join(Config.CYBER_MAP_DIR,f'cyber_map_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png')
            plt.savefig(fp,dpi=Config.VIZ_DPI,bbox_inches='tight'); plt.close()
            return fp
        except Exception as e:
            logger.error(f"3D map error: {e}"); return None

    def _get_threat_location(self,t):
        return self.country_codes.get(self._guess_country(t),(0,0))

    def _guess_country(self,t):
        url=t.get('url','').lower()
        for tld,c in {'.cn':'CN','.ru':'RU','.br':'BR','.in':'IN','.uk':'UK',
                      '.de':'DE','.fr':'FR','.ca':'CA','.au':'AU','.jp':'JP'}.items():
            if tld in url: return c
        return 'US' if any(x in url for x in ['.com','.net','.org']) else 'Unknown'


class AdversarialDefense:
    def __init__(self):
        self.attack_history=deque(maxlen=1000); self.evasion_attempts_detected=0

    def generate_adversarial_samples(self,X:pd.DataFrame,y:np.ndarray,epsilon:float=None)->pd.DataFrame:
        if epsilon is None: epsilon=Config.ADVERSARIAL_NOISE_EPSILON
        X_adv=X.copy()
        for i in range(len(X_adv)):
            noise=np.random.normal(0,epsilon,X_adv.shape[1])
            X_adv.iloc[i]=(X_adv.iloc[i]+noise).clip(X.min(),X.max())
        return X_adv

    def train_adversarial_robust_model(self,X_train,y_train,X_test,y_test):
        try:
            n=int(len(X_train)*Config.ADVERSARIAL_TRAINING_RATIO)
            idx=np.random.choice(len(X_train),n,replace=False)
            Xa=self.generate_adversarial_samples(X_train.iloc[idx],y_train[idx])
            Xc=pd.concat([X_train,Xa],ignore_index=True)
            yc=np.concatenate([y_train,y_train[idx]])
            m=RandomForestClassifier(n_estimators=300,max_depth=25,min_samples_split=5,
                                     min_samples_leaf=2,random_state=Config.RANDOM_STATE,n_jobs=-1)
            m.fit(Xc,yc)
            return m,{'clean_accuracy':m.score(X_test,y_test)}
        except Exception as e:
            logger.error(f"Adversarial training failed: {e}")
            m=RandomForestClassifier(n_estimators=200,random_state=Config.RANDOM_STATE)
            m.fit(X_train,y_train)
            return m,{'clean_accuracy':m.score(X_test,y_test)}

    def detect_evasion_attempt(self,features:Dict,model_predictions:Dict)->Dict:
        score=0; reasons=[]
        try:
            fv=[v for v in features.values() if isinstance(v,(int,float))]
            if fv and len(fv)>1 and np.var(fv)>15: score+=15; reasons.append("High feature variance")
            if len(model_predictions)>1:
                pv=[v for v in model_predictions.values() if v!=0]
                if pv and len(pv)>1 and np.var(pv)>0.2: score+=20; reasons.append("Model disagreement")
            is_ev=score>=(Config.EVASION_THRESHOLD*100)
            if is_ev: self.evasion_attempts_detected+=1
            return {'is_evasion':is_ev,'evasion_score':score,'reasons':reasons}
        except:
            return {'is_evasion':False,'evasion_score':0,'reasons':[]}

    def train_evasion_detector(self,X_train,y_train,X_test,y_test)->Dict:
        try:
            det=IsolationForest(n_estimators=150,contamination=0.1,random_state=Config.RANDOM_STATE,n_jobs=-1)
            det.fit(X_train[y_train==0])
            return {'if':det}
        except Exception as e:
            logger.error(f"Evasion detector training failed: {e}"); return {}


class ThreatIntelligenceFusion:
    def __init__(self):
        self.openphish_urls=set(); self.urlhaus_urls=set(); self.last_update=None
        self.load_threat_feeds()

    def load_threat_feeds(self):
        Colors.print_info("Loading threat intelligence feeds...")
        try:
            r=requests.get("https://openphish.com/feed.txt",timeout=10)
            if r.status_code==200:
                self.openphish_urls=set([u.strip() for u in r.text.strip().split('\n') if u.strip()][:1000])
                Colors.print_success(f"✓ OpenPhish: {len(self.openphish_urls)} threats")
        except Exception as e:
            Colors.print_warning(f"OpenPhish load failed: {str(e)[:50]}"); self.openphish_urls=set()
        try:
            r=requests.get("https://urlhaus.abuse.ch/downloads/csv_recent/",timeout=10)
            if r.status_code==200:
                urls=[]
                for line in r.text.strip().split('\n')[9:]:
                    try:
                        p=line.split('","')
                        if len(p)>2: urls.append(p[2].strip('"'))
                    except: continue
                self.urlhaus_urls=set(urls[:1000])
                Colors.print_success(f"✓ URLhaus: {len(self.urlhaus_urls)} threats")
        except Exception as e:
            Colors.print_warning(f"URLhaus load failed: {str(e)[:50]}"); self.urlhaus_urls=set()
        self.last_update=datetime.now()

    def check_url_reputation(self,url:str)->Dict:
        score=0; matches=[]
        try:
            ul=url.lower().strip()
            if any(ul in t.lower() or t.lower() in ul for t in self.openphish_urls): score=80; matches.append('OpenPhish')
            if any(ul in t.lower() or t.lower() in ul for t in self.urlhaus_urls): score=max(score,75); matches.append('URLhaus')
        except: pass
        return {'combined_risk_score':score,'threats_found':matches}


class MultiModelEnsemble:
    def __init__(self):
        self.models={}; self.scalers={}; self.feature_names=[]
        self.adversarial_defense=AdversarialDefense()

    def train_ensemble(self,X_train,y_train,X_test,y_test)->Dict:
        Colors.print_info("Training ensemble models...")
        self.models['rf']=RandomForestClassifier(n_estimators=200,max_depth=20,min_samples_split=5,
                                                  min_samples_leaf=2,random_state=Config.RANDOM_STATE,n_jobs=-1)
        self.models['rf'].fit(X_train,y_train)
        rf=self.models['rf'].score(X_test,y_test)
        Colors.print_success(f"✓ Random Forest: {rf:.2%}")

        self.models['gb']=GradientBoostingClassifier(n_estimators=200,learning_rate=0.1,max_depth=5,random_state=Config.RANDOM_STATE)
        self.models['gb'].fit(X_train,y_train)
        gb=self.models['gb'].score(X_test,y_test)
        Colors.print_success(f"✓ Gradient Boosting: {gb:.2%}")

        if Config.ADVERSARIAL_ENABLED:
            Xtr=pd.DataFrame(X_train,columns=self.feature_names)
            Xte=pd.DataFrame(X_test,columns=self.feature_names)
            self.models['adversarial_rf'],ar=self.adversarial_defense.train_adversarial_robust_model(Xtr,y_train,Xte,y_test)
            Colors.print_success(f"✓ Adversarial Model: {ar['clean_accuracy']:.2%}")

        Xtr=pd.DataFrame(X_train,columns=self.feature_names)
        Xte=pd.DataFrame(X_test,columns=self.feature_names)
        self.models['evasion_detector']=self.adversarial_defense.train_evasion_detector(Xtr,y_train,Xte,y_test)
        return {'rf_score':rf,'gb_score':gb}

    def predict_ensemble(self,features:Dict,url:str,threat_intel_score:float=0,threat_dna_score:float=0)->Dict:
        try:
            X=pd.DataFrame([features]).reindex(columns=self.feature_names,fill_value=0)
            Xs=self.scalers['standard'].transform(X)
            preds={}
            if 'rf'             in self.models: preds['rf']             =self.models['rf'].predict_proba(Xs)[0][1]
            if 'gb'             in self.models: preds['gb']             =self.models['gb'].predict_proba(Xs)[0][1]
            if 'adversarial_rf' in self.models: preds['adversarial_rf']=self.models['adversarial_rf'].predict_proba(Xs)[0][1]
            preds['threat_intel']=threat_intel_score/100.0
            preds['threat_dna']=threat_dna_score/100.0
            ev=self.adversarial_defense.detect_evasion_attempt(features,preds)
            preds['evasion_detector']=ev['evasion_score']/100.0
            tw=sum(Config.WEIGHTS.get(k,0.1) for k in preds)
            ws=sum(p*Config.WEIGHTS.get(k,0.1) for k,p in preds.items())
            fp=ws/tw if tw>0 else 0.5
            if ev['is_evasion']: fp=min(fp*1.3,1.0)
            return {'final_probability':fp,'is_phishing':fp>0.5,'model_predictions':preds,
                    'confidence':max(fp,1-fp),'evasion_detection':ev}
        except Exception as e:
            logger.error(f"Ensemble prediction error: {e}")
            return {'final_probability':0.5,'is_phishing':False,'model_predictions':{},
                    'confidence':0.5,'evasion_detection':{'is_evasion':False,'evasion_score':0,'reasons':[]}}


# ==================== FEATURE EXTRACTION ====================

def calculate_entropy(text:str)->float:
    if not text: return 0
    entropy=0
    for c in set(text):
        p=text.count(c)/len(text)
        entropy-=p*np.log2(p)
    return entropy


def extract_features(url:str)->Tuple[Optional[Dict],Optional[str]]:
    try:
        if not url.startswith(('http://','https://')): url='https://'+url
        parsed=urlparse(url)
        domain=parsed.netloc.lower(); path=parsed.path.lower(); ul=url.lower()
        suspicious_keywords=['verify','account','update','confirm','login','secure','banking',
                             'paypal','amazon','netflix','suspended','locked','signin','webscr','password','credential']
        suspicious_tlds=['.xyz','.top','.win','.club','.online','.site','.tk','.ml','.ga','.cf','.gq','.pw','.cc']
        brand_names=['paypal','amazon','netflix','linkedin','facebook','google','microsoft','apple','ebay','instagram']
        kc=sum(1 for kw in suspicious_keywords if kw in ul)
        st=any(t in domain for t in suspicious_tlds)
        hb=any(b in domain for b in brand_names)
        ht=any(c in domain for c in ['1','0','l']) and hb
        he=any(e in path for e in ['.exe','.scr','.bat','.cmd','.zip'])
        return {
            'url_length':   min(len(url)/100.0,2.0),
            'domain_length':min(len(domain)/50.0,2.0),
            'subdomain_count':min(max(0,domain.count('.')-1),5),
            'has_https':    1 if url.startswith('https://') else 0,
            'has_ip':       2 if re.match(r'(\d+\.){3}\d+',domain) else 0,
            'login_word':   2 if 'login' in ul else 0,
            'entropy':      min(calculate_entropy(url)/5.0,2.0),
            'keyword_count':min(kc,5),
            'has_suspicious_tld':3 if st else 0,
            'has_brand_typo':   3 if ht else 0,
            'has_executable':   5 if he else 0,
            'multiple_subdomains':2 if len(domain.split('.'))>3 else 0,
            'has_hyphen':   1 if '-' in domain else 0,
            'has_numbers':  1 if any(c.isdigit() for c in domain) else 0,
            'has_at_symbol':3 if '@' in url else 0,
            'suspicious_pattern':2 if (hb and not domain.endswith(('.com','.org','.net'))) else 0
        }, url
    except Exception as e:
        logger.error(f"Feature extraction failed: {e}"); return None,None


def load_data(file_path:str)->pd.DataFrame:
    try:
        Colors.print_info(f"Loading dataset from: {file_path}")
        df=pd.read_csv(file_path,encoding='utf-8',on_bad_lines='skip')
        uc=lc=None
        for col in df.columns:
            cl=col.lower()
            if cl in ['url','text','link','website']: uc=col
            if cl in ['label','class','target','type']: lc=col
        if uc is None or lc is None: raise ValueError("Could not find URL or label columns")
        dc=df[[uc,lc]].copy(); dc.columns=['url','label']; dc=dc.dropna()
        dc['label']=pd.to_numeric(dc['label'],errors='coerce').astype(int)
        dc=dc[dc['label'].isin([0,1])]
        Colors.print_success(f"✓ Loaded {len(dc)} records")
        Colors.print_info(f"  Phishing: {sum(dc['label']==1)} | Legitimate: {sum(dc['label']==0)}")
        return dc
    except Exception as e:
        logger.error(f"Data loading failed: {e}"); raise


def train_model(df:pd.DataFrame)->Dict:
    Colors.print_header("TRAINING MODEL PIPELINE")
    Colors.print_info("Extracting features...")
    fl=[]; labels=[]
    for _,row in tqdm(df.iterrows(),total=len(df),desc="Processing URLs"):
        f,_=extract_features(row['url'])
        if f: fl.append(f); labels.append(row['label'])
    X=pd.DataFrame(fl).fillna(0); y=np.array(labels)
    Colors.print_success(f"✓ {len(X)} feature vectors")
    Xtr,Xte,ytr,yte=train_test_split(X,y,test_size=Config.TEST_SIZE,random_state=Config.RANDOM_STATE,stratify=y)
    Colors.print_info(f"Train: {len(Xtr)} | Test: {len(Xte)}")
    scaler=StandardScaler()
    Xtrs=scaler.fit_transform(Xtr); Xtes=scaler.transform(Xte)
    ens=MultiModelEnsemble(); ens.feature_names=list(X.columns); ens.scalers['standard']=scaler
    ens.train_ensemble(Xtrs,ytr,Xtes,yte)
    preds=[ens.predict_ensemble(Xte.iloc[i].to_dict(),'',0,0)['is_phishing'] for i in range(len(Xte))]
    acc=accuracy_score(yte,preds)
    Colors.print_success(f"✓ Ensemble Accuracy: {acc:.2%}")
    md={'ensemble':ens,'scaler':scaler,'feature_names':list(X.columns),'accuracy':acc}
    import joblib; joblib.dump(md,Config.MODEL_FILE)
    Colors.print_success(f"✓ Model saved: {Config.MODEL_FILE}")
    return md


def load_model_safe()->Optional[Dict]:
    """
    Load the saved model, automatically deleting and retraining
    if the file is stale / built with an incompatible sklearn version.
    """
    import joblib
    if not os.path.exists(Config.MODEL_FILE):
        return None
    try:
        Colors.print_info("Loading trained model...")
        md = joblib.load(Config.MODEL_FILE)
        Colors.print_success(f"✓ Model loaded (Accuracy: {md['accuracy']:.2%})")
        return md
    except Exception as e:
        Colors.print_warning(
            f"Saved model is incompatible with current sklearn "
            f"({type(e).__name__}: {str(e)[:80]})"
        )
        Colors.print_warning("Deleting stale model and retraining from scratch...")
        try:
            os.remove(Config.MODEL_FILE)
            Colors.print_success("✓ Stale model deleted")
        except Exception as de:
            Colors.print_error(f"Could not delete model file: {de}")
        return None


# ==================== PREDICTION ====================

def predict_url_with_defenses(url:str, model_data:Dict,
                               threat_intel:ThreatIntelligenceFusion,
                               threat_dna_score:float=0)->Dict:
    features,nurl=extract_features(url)
    if not features: return {'error':'Feature extraction failed'}
    try:
        ens=model_data['ensemble']
        ti=threat_intel.check_url_reputation(nurl)
        ts=ti['combined_risk_score']
        er=ens.predict_ensemble(features,nurl,ts,threat_dna_score)
        ml_risk=int(er['final_probability']*100)

        hs=0; ul=nurl.lower()
        if features.get('has_executable',0)>0:     hs=max(hs,85)
        if features.get('has_brand_typo',0)>0:     hs=max(hs,80)
        if features.get('has_suspicious_tld',0)>0: hs=max(hs,75)
        kc=features.get('keyword_count',0)
        if kc>=4:   hs=max(hs,75)
        elif kc>=3: hs=max(hs,65)
        elif kc>=2: hs=max(hs,55)
        if 'verify' in ul and 'account' in ul: hs=max(hs,70)

        if   hs>=75: base=int(ml_risk*0.15+hs*0.85)
        elif hs>=60: base=int(ml_risk*0.25+hs*0.75)
        elif hs>0:   base=int(ml_risk*0.5+hs*0.5)
        else:        base=ml_risk

        if ts>0: base=min(int(base*0.6+ts*0.4),100)
        # Clamp to valid 0-100 integer
        base=max(0,min(int(base),100))

        if base>=Config.HIGH_RISK:
            rl="HIGH RISK";   rec="⛔ DO NOT OPEN - High probability of phishing"; ph=True
        elif base>=Config.MEDIUM_RISK:
            rl="MEDIUM RISK"; rec="⚠️ VERIFY FIRST - Exercise caution";            ph=True
        else:
            rl="SAFE";        rec="✅ Appears legitimate";                          ph=False

        return {'url':nurl,'is_phishing':ph,'risk_score':base,'risk_level':rl,
                'recommendation':rec,'confidence':round(er['confidence'],4),
                'model_predictions':er['model_predictions'],'evasion_detection':er['evasion_detection'],
                'threat_intelligence':ti,'heuristic_score':hs,'ml_risk':ml_risk}
    except Exception as e:
        logger.error(f"Prediction error: {e}"); return {'error':f'Prediction failed: {e}'}


# ==================== BROWSER BLOCKING ====================

def safe_open_url(url:str, result:Dict)->bool:
    """
    Open URL in browser ONLY when risk < MEDIUM_RISK (safe).
    Any suspicious or phishing URL is BLOCKED from opening.
    """
    risk=result.get('risk_score',100)
    is_phishing=result.get('is_phishing',True)

    if is_phishing or risk>=Config.MEDIUM_RISK:
        print(f"\n{Colors.FAIL}{'='*62}")
        print(f"  🚫  URL BLOCKED — NOT opened in browser")
        print(f"  Risk Score : {risk}/100  |  Level: {result.get('risk_level','UNKNOWN')}")
        print(f"  The URL was fully analyzed but opening was prevented")
        print(f"  to protect you from potential phishing or malware.")
        print(f"{'='*62}{Colors.ENDC}")
        return False
    else:
        try:
            Colors.print_success("URL is safe — opening in browser...")
            webbrowser.open(url)
            Colors.print_success("✓ Opened successfully")
            return True
        except Exception as e:
            Colors.print_error(f"Failed to open URL: {e}")
            return False


# ==================== DISPLAY ====================

def display_result(result:Dict, show_visualization:bool=True):
    Colors.print_header("🛡️ ANALYSIS RESULT")
    print(f"\n{Colors.BOLD}URL:{Colors.ENDC} {result['url'][:70]}...")

    sc=Colors.FAIL if result['is_phishing'] else Colors.OKGREEN
    st="⚠️  PHISHING DETECTED" if result['is_phishing'] else "✅  LEGITIMATE"
    print(f"\n{Colors.BOLD}Status:{Colors.ENDC} {sc}{st}{Colors.ENDC}")

    risk=result['risk_score']
    filled=int(50*risk/100)
    bc=Colors.FAIL if risk>=Config.HIGH_RISK else Colors.WARNING if risk>=Config.MEDIUM_RISK else Colors.OKGREEN
    bar='█'*filled+'░'*(50-filled)
    print(f"\n{Colors.BOLD}Risk Score:{Colors.ENDC} {bc}{bar}{Colors.ENDC} {risk}/100")
    print(f"{Colors.BOLD}Risk Level:{Colors.ENDC} {result['risk_level']}")
    print(f"{Colors.BOLD}Recommendation:{Colors.ENDC} {result['recommendation']}")

    if result.get('threat_intelligence',{}).get('threats_found'):
        print(f"\n{Colors.FAIL}⚠️  Matched threat feeds: {', '.join(result['threat_intelligence']['threats_found'])}{Colors.ENDC}")
    if result.get('evasion_detection',{}).get('is_evasion'):
        print(f"{Colors.WARNING}⚠️  Evasion attempt detected!{Colors.ENDC}")

    if show_visualization:
        try:
            vp=generate_url_visualization(result)
            if vp:
                Colors.print_success(f"✓ Visualization saved: {os.path.basename(vp)}")
                try: webbrowser.open(f'file://{os.path.abspath(vp)}'); Colors.print_success("✓ Visualization opened")
                except: pass
        except Exception as e:
            logger.warning(f"Visualization failed: {e}")

    safe_open_url(result['url'], result)
    print()


def generate_url_visualization(result:Dict)->Optional[str]:
    try:
        fig,axes=plt.subplots(2,2,figsize=(14,10))
        risk=result['risk_score']

        ax1=axes[0,0]
        ax1.pie([Config.MEDIUM_RISK,Config.HIGH_RISK-Config.MEDIUM_RISK,100-Config.HIGH_RISK],
                colors=['green','orange','red'],startangle=180,counterclock=False)
        angle=180-(risk*1.8)
        ax1.annotate('',xy=(0.6*np.cos(np.radians(angle)),0.6*np.sin(np.radians(angle))),
                    xytext=(0,0),arrowprops=dict(arrowstyle='->',lw=3,color='black'))
        ax1.set_title(f'Risk Score: {risk}/100',fontweight='bold',fontsize=12)

        ax2=axes[0,1]
        if result.get('model_predictions'):
            preds=result['model_predictions']
            vals=[v*100 for v in preds.values()]
            ax2.barh(list(preds.keys()),vals,
                     color=['green' if v<50 else 'orange' if v<70 else 'red' for v in vals],alpha=0.7)
            ax2.set_xlabel('Confidence (%)'); ax2.set_title('Model Predictions',fontweight='bold')
            ax2.axvline(x=50,color='black',linestyle='--',alpha=0.5); ax2.grid(axis='x',alpha=0.3)

        ax3=axes[1,0]
        fs={'Heuristic':result.get('heuristic_score',0),'ML Model':result.get('ml_risk',0),
            'Threat Intel':result.get('threat_intelligence',{}).get('combined_risk_score',0)}
        fv=list(fs.values())
        ax3.bar(list(fs.keys()),fv,color=['green' if v<45 else 'orange' if v<70 else 'red' for v in fv],alpha=0.7)
        ax3.set_ylabel('Risk Score'); ax3.set_title('Risk Components',fontweight='bold')
        ax3.axhline(y=Config.HIGH_RISK,color='red',linestyle='--',alpha=0.5,label='High Risk')
        ax3.axhline(y=Config.MEDIUM_RISK,color='orange',linestyle='--',alpha=0.5,label='Medium Risk')
        ax3.legend(); ax3.grid(axis='y',alpha=0.3)

        ax4=axes[1,1]; ax4.axis('off')
        ax4.text(0.1,0.5,
                 f"URL ANALYSIS SUMMARY\n\nURL: {result['url'][:60]}...\n\n"
                 f"Status:     {result['risk_level']}\nRisk Score: {risk}/100\n"
                 f"Confidence: {result.get('confidence',0):.1%}\n\n"
                 f"Recommendation:\n{result['recommendation']}\n\n"
                 f"Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}",
                 fontsize=10,verticalalignment='center',fontfamily='monospace',
                 bbox=dict(boxstyle='round',facecolor='wheat',alpha=0.3))

        plt.suptitle(f'PHISHING ANALYSIS - {result["risk_level"]}',fontsize=14,fontweight='bold')
        plt.tight_layout()
        vf=os.path.join(Config.VISUALIZATION_DIR,f'url_analysis_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png')
        plt.savefig(vf,dpi=Config.VIZ_DPI,bbox_inches='tight'); plt.close()
        return vf
    except Exception as e:
        logger.error(f"Visualization failed: {e}"); return None


# ==================== MAIN ====================

def print_menu():
    print(f"\n{Colors.BOLD}{'='*80}")
    print(f"  🧬 PHISHING DETECTION v10.0 - COMMAND CENTER")
    print(f"{'='*80}{Colors.ENDC}")

    print(f"\n{Colors.HEADER}═══ CORE ANALYSIS ═══{Colors.ENDC}")
    print(f"{Colors.OKCYAN}[1]{Colors.ENDC}  🔗 Quick URL Scan")
    print(f"{Colors.OKCYAN}[2]{Colors.ENDC}  🛡️  Full Analysis with Report")

    print(f"\n{Colors.HEADER}═══ UTILITIES ═══{Colors.ENDC}")
    print(f"{Colors.OKCYAN}[3]{Colors.ENDC}  📊 Threat Intel Stats")
    print(f"{Colors.OKCYAN}[4]{Colors.ENDC}  🔄 Reload Threat Feeds")
    print(f"{Colors.OKCYAN}[5]{Colors.ENDC}  ❓ Help & Information")

    print(f"\n{Colors.HEADER}═══ ADVANCED FEATURES ═══{Colors.ENDC}")
    print(f"{Colors.OKGREEN}[6]{Colors.ENDC}  🌍 3D Cyber Map")
    print(f"{Colors.OKGREEN}[7]{Colors.ENDC}  📊 Real-Time Dashboard")
    print(f"{Colors.OKGREEN}[8]{Colors.ENDC}  🤖 AI Chat Assistant")
    print(f"{Colors.OKGREEN}[9]{Colors.ENDC}  📈 Analytics & Trends")

    print(f"\n{Colors.HEADER}═══ 🧬 THREAT DNA ═══{Colors.ENDC}")
    print(f"{Colors.OKGREEN}[10]{Colors.ENDC} 🕷️  Deep Crawl & DNA Collection")
    print(f"{Colors.OKGREEN}[11]{Colors.ENDC} 🔍 Find Similar Threats")
    print(f"{Colors.OKGREEN}[12]{Colors.ENDC} 🎯 Campaign Detection")
    print(f"{Colors.OKGREEN}[13]{Colors.ENDC} 📊 DNA Database Stats")

    print(f"\n{Colors.HEADER}═══ EXIT ═══{Colors.ENDC}")
    print(f"{Colors.OKCYAN}[0]{Colors.ENDC}  🚪 Exit")
    print(f"{Colors.BOLD}{'='*80}{Colors.ENDC}")


def main():
    Colors.print_header("🧬 PHISHING DETECTION SYSTEM v10.0")
    print(f"{Colors.OKCYAN}Phishing URLs blocked from browser  |  Sequential menu  |  Score 0-100{Colors.ENDC}\n")

    if not Config.DATA_FILE:
        Colors.print_error("Dataset not found!")
        for p in Config.POSSIBLE_DATA_PATHS: print(f"  • {p}")
        return

    Colors.print_info("Initializing systems...")
    threat_intel    = ThreatIntelligenceFusion()
    monitor         = RealTimeMonitor()
    cyber_map       = CyberMapVisualizer()
    chat_assistant  = SecurityChatAssistant()
    crawler         = ThreatDNACrawler()        if HAS_PLAYWRIGHT   else None
    embedding_gen   = EmbeddingGenerator()      if HAS_TRANSFORMERS else None
    vector_db       = ThreatVectorDB()          if HAS_VECTORDB     else None
    campaign_det    = CampaignDetector()

    print(f"\n{Colors.BOLD}Capabilities:{Colors.ENDC}")
    print(f"  Playwright:   {'✓' if HAS_PLAYWRIGHT  else '✗'}  Transformers: {'✓' if HAS_TRANSFORMERS else '✗'}"
          f"  VectorDB: {'✓' if HAS_VECTORDB else '✗'}  OpenCV: {'✓' if HAS_CV else '✗'}")

    try:
        # ── Safe model loading with automatic stale-model recovery ──
        model_data = load_model_safe()
        if model_data is None:
            Colors.print_warning("No valid model found — training now...")
            model_data = train_model(load_data(Config.DATA_FILE))

        monitor.start_monitoring()
        collected_dnas=[]

        while True:
            print_menu()
            choice=input(f"\n{Colors.BOLD}Select option:{Colors.ENDC} ").strip()

            # ── EXIT ──
            if choice=='0':
                monitor.stop_monitoring()
                Colors.print_success("\n👋 Stay safe! Goodbye!")
                break

            # ── 1: Quick URL Scan ──
            elif choice=='1':
                url=input("\n🔗 Enter URL: ").strip()
                if not url: continue
                result=predict_url_with_defenses(url,model_data,threat_intel)
                if result.get('error'):
                    Colors.print_error(result['error'])
                else:
                    monitor.log_detection(result)
                    display_result(result,show_visualization=True)
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 2: Full Analysis with Report ──
            elif choice=='2':
                url=input("\n🔗 Enter URL: ").strip()
                if not url: continue
                Colors.print_info("Running full analysis...")
                result=predict_url_with_defenses(url,model_data,threat_intel)
                if result.get('error'):
                    Colors.print_error(result['error'])
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}"); continue
                monitor.log_detection(result)
                display_result(result,show_visualization=True)
                if input(f"\n{Colors.BOLD}Save JSON report? (y/n):{Colors.ENDC} ").strip().lower()=='y':
                    rp=os.path.join(Config.REPORTS_DIR,f"report_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json")
                    with open(rp,'w') as f: json.dump(result,f,indent=2,default=str)
                    Colors.print_success(f"✓ Report saved: {os.path.basename(rp)}")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 3: Threat Intel Stats ──
            elif choice=='3':
                Colors.print_header("📊 THREAT INTELLIGENCE STATISTICS")
                print(f"\n  OpenPhish URLs : {len(threat_intel.openphish_urls)}")
                print(f"  URLhaus URLs   : {len(threat_intel.urlhaus_urls)}")
                if threat_intel.last_update:
                    print(f"  Last Update    : {threat_intel.last_update.strftime('%Y-%m-%d %H:%M:%S')}")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 4: Reload Threat Feeds ──
            elif choice=='4':
                Colors.print_info("Reloading threat intelligence feeds...")
                threat_intel.load_threat_feeds()
                Colors.print_success("✓ Feeds reloaded!")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 5: Help ──
            elif choice=='5':
                Colors.print_header("❓ HELP & INFORMATION")
                print(f"\n{Colors.BOLD}🚫 URL BROWSER BLOCKING:{Colors.ENDC}")
                print("  Suspicious/phishing URLs (risk ≥ 45) are blocked from opening.")
                print("  Only URLs with risk < 45 are opened in the browser.\n")
                print(f"{Colors.BOLD}📊 RISK SCORE:{Colors.ENDC}")
                print("  0 – 44  : SAFE      (green)  — may open in browser")
                print("  45 – 69 : MEDIUM    (orange) — BLOCKED")
                print("  70 –100 : HIGH RISK (red)    — BLOCKED\n")
                print(f"{Colors.BOLD}🧬 THREAT DNA:{Colors.ENDC}")
                print("  Deep web crawling, multi-modal embeddings, vector similarity.\n")
                print(f"{Colors.BOLD}🌐 THREAT INTELLIGENCE:{Colors.ENDC}")
                print("  OpenPhish + URLhaus live feed integration.\n")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 6: 3D Cyber Map ──
            elif choice=='6':
                Colors.print_header("🌍 3D CYBER MAP")
                if not monitor.detection_history:
                    Colors.print_warning("No data yet. Analyze some URLs first!")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}"); continue
                mp=cyber_map.create_3d_cyber_map(list(monitor.detection_history))
                if mp:
                    Colors.print_success(f"✓ Map saved: {os.path.basename(mp)}")
                    try: webbrowser.open(f'file://{os.path.abspath(mp)}')
                    except: Colors.print_info(f"Path: {mp}")
                else:
                    Colors.print_error("Map generation failed")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 7: Real-Time Dashboard ──
            elif choice=='7':
                Colors.print_header("📊 REAL-TIME DASHBOARD")
                s=monitor.get_current_stats()
                print(f"\n  Total Scans:       {s['total_scans']}")
                print(f"  Phishing Detected: {s['phishing_detected']}")
                print(f"  Safe URLs:         {s['safe_urls']}")
                print(f"  Detection Rate:    {s['detection_rate']:.1f}%")
                print(f"\n  CPU:    {s['cpu_usage']:.1f}%")
                print(f"  Memory: {s['memory_usage']:.1f}%")
                print(f"  Active Threats: {s['active_threats']}")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 8: AI Chat Assistant ──
            elif choice=='8':
                Colors.print_header("🤖 AI SECURITY CHAT ASSISTANT")
                print(f"{Colors.OKCYAN}Type 'back' to return.{Colors.ENDC}\n")
                chat_assistant.update_context(
                    detections=list(monitor.detection_history),
                    system_health={
                        'cpu':   monitor.system_metrics[-1]['cpu_percent']    if monitor.system_metrics else 0,
                        'memory':monitor.system_metrics[-1]['memory_percent'] if monitor.system_metrics else 0
                    }
                )
                while True:
                    ui=input(f"\n{Colors.OKBLUE}You:{Colors.ENDC} ").strip()
                    if ui.lower()=='back': break
                    if not ui: continue
                    print(f"\n{Colors.OKGREEN}Assistant:{Colors.ENDC} {chat_assistant.process_query(ui)}")

            # ── 9: Analytics & Trends ──
            elif choice=='9':
                Colors.print_header("📈 ADVANCED ANALYTICS")
                if not monitor.detection_history:
                    Colors.print_warning("No data yet."); input(f"\n{Colors.BOLD}Press Enter...{Colors.ENDC}"); continue
                det=list(monitor.detection_history); tot=len(det)
                phi=sum(1 for d in det if d['is_phishing'])
                print(f"\n  Analyzed:   {tot}")
                print(f"  Phishing:   {phi} ({phi/tot*100:.1f}%)")
                print(f"  Safe:       {tot-phi}")
                hi=sum(1 for d in det if d['risk_score']>=Config.HIGH_RISK)
                me=sum(1 for d in det if Config.MEDIUM_RISK<=d['risk_score']<Config.HIGH_RISK)
                lo=tot-hi-me
                print(f"\n  {Colors.FAIL}High:   {hi:3d} ({hi/tot*100:5.1f}%){Colors.ENDC}")
                print(f"  {Colors.WARNING}Medium: {me:3d} ({me/tot*100:5.1f}%){Colors.ENDC}")
                print(f"  {Colors.OKGREEN}Low:    {lo:3d} ({lo/tot*100:5.1f}%){Colors.ENDC}")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 10: Deep Crawl & DNA ──
            elif choice=='10':
                Colors.print_header("🕷️ DEEP CRAWL & DNA COLLECTION")
                if not crawler:
                    Colors.print_error("Playwright not installed!")
                    print("  pip install playwright && playwright install chromium")
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}"); continue
                url=input("\n🔗 Enter URL to crawl: ").strip()
                if not url: continue
                Colors.print_info("Starting deep crawl (30-60 seconds)...")
                dna=crawler.crawl_url_sync(url)
                if not dna:
                    Colors.print_warning("Crawl failed — running standard analysis...")
                    result=predict_url_with_defenses(url,model_data,threat_intel)
                    if not result.get('error'):
                        monitor.log_detection(result); display_result(result)
                    input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}"); continue
                if embedding_gen:
                    Colors.print_info("Generating embeddings...")
                    dna=embedding_gen.enrich_threat_dna(dna)
                    Colors.print_success("✓ Embeddings generated" if dna.combined_embedding is not None else "⚠ Partial embeddings")
                result=predict_url_with_defenses(url,model_data,threat_intel)
                if not result.get('error'):
                    dna.risk_score=result['risk_score']; dna.threat_category=result['risk_level']
                    monitor.log_detection(result)
                    try:
                        vp=generate_url_visualization(result)
                        if vp:
                            try: webbrowser.open(f'file://{os.path.abspath(vp)}')
                            except: pass
                    except: pass
                if vector_db and dna.combined_embedding is not None:
                    if vector_db.add_threat(dna): Colors.print_success("✓ Stored in vector DB")
                collected_dnas.append(dna)
                Colors.print_header("🧬 THREAT DNA ANALYSIS")
                print(f"  Threat ID   : {dna.threat_id}")
                print(f"  Risk Score  : {dna.risk_score}/100")
                print(f"  Category    : {dna.threat_category or 'N/A'}")
                print(f"  Screenshot  : {'✓' if dna.screenshot_path else '✗'}")
                print(f"  HTML Dump   : {'✓' if dna.html_content   else '✗'}")
                print(f"  Links Found : {len(dna.extracted_links)}")
                print(f"  Forms Found : {len(dna.extracted_forms)}")
                print(f"  Fingerprint : {dna.get_fingerprint()}")
                dp=os.path.join(Config.THREAT_DNA_DIR,f"threat_dna_{dna.threat_id}.json")
                with open(dp,'w',encoding='utf-8') as f: json.dump(dna.to_dict(),f,indent=2,ensure_ascii=False)
                Colors.print_success(f"✓ DNA saved: threat_dna_{dna.threat_id}.json")
                if not result.get('error'): safe_open_url(url,result)
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 11: Find Similar Threats ──
            elif choice=='11':
                Colors.print_header("🔍 FIND SIMILAR THREATS")
                if not vector_db:   Colors.print_error("Vector DB not available!"); input(f"\n{Colors.BOLD}Press Enter...{Colors.ENDC}"); continue
                if not collected_dnas: Colors.print_warning("No DNA samples yet. Use [10] first!"); input(f"\n{Colors.BOLD}Press Enter...{Colors.ENDC}"); continue
                avail=collected_dnas[-10:]
                for i,d in enumerate(avail,1): print(f"  [{i}] {d.threat_id}  Risk:{d.risk_score}/100  {d.url[:50]}...")
                try:
                    idx=int(input(f"\nSelect (1-{len(avail)}): ").strip())
                    if not 1<=idx<=len(avail): raise ValueError()
                    sel=avail[idx-1]
                    if sel.combined_embedding is None: Colors.print_error("No embedding available!"); input(f"\n{Colors.BOLD}Press Enter...{Colors.ENDC}"); continue
                    similar=vector_db.find_similar_threats(sel,top_k=5)
                    if not similar:
                        Colors.print_warning("No similar threats found.")
                    else:
                        print(f"\n{Colors.BOLD}━━━ SIMILAR THREATS ━━━{Colors.ENDC}")
                        for i,s in enumerate(similar,1):
                            c=Colors.FAIL if s['similarity']>0.9 else Colors.WARNING if s['similarity']>0.8 else Colors.OKGREEN
                            print(f"{c}[{i}] Similarity: {s['similarity']:.1%}{Colors.ENDC}  Risk: {s['metadata'].get('risk_score',0)}/100")
                            print(f"    URL: {s['metadata'].get('url','N/A')[:70]}...")
                        if similar[0]['similarity']>0.9:
                            print(f"{Colors.FAIL}⚠️  Very high similarity — possible campaign!{Colors.ENDC}")
                except (ValueError, IndexError):
                    Colors.print_error("Invalid input.")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 12: Campaign Detection ──
            elif choice=='12':
                Colors.print_header("🎯 CAMPAIGN DETECTION")
                if not collected_dnas: Colors.print_warning("No DNA samples yet. Use [10] first!"); input(f"\n{Colors.BOLD}Press Enter...{Colors.ENDC}"); continue
                if len(collected_dnas)<Config.CAMPAIGN_MIN_SAMPLES:
                    Colors.print_warning(f"Need ≥ {Config.CAMPAIGN_MIN_SAMPLES} samples. Have: {len(collected_dnas)}")
                    input(f"\n{Colors.BOLD}Press Enter...{Colors.ENDC}"); continue
                campaigns=campaign_det.detect_campaigns(collected_dnas)
                print(f"\n  Campaigns:   {len([c for c in campaigns if c!='UNCLUSTERED'])}")
                print(f"  Unclustered: {len(campaigns.get('UNCLUSTERED',[]))}")
                for cid,threats in campaigns.items():
                    if not threats: continue
                    avg=np.mean([t.risk_score for t in threats])
                    col=Colors.FAIL if avg>70 else Colors.WARNING if avg>45 else Colors.OKGREEN
                    if cid=='UNCLUSTERED':
                        print(f"\n{Colors.BOLD}UNCLUSTERED ({len(threats)}){Colors.ENDC}")
                    else:
                        print(f"\n{col}{cid}{Colors.ENDC} ({len(threats)} threats | avg risk: {avg:.1f}/100)")
                    for t in threats[:3]: print(f"    • {t.url[:60]}... ({t.risk_score}/100)")
                    if len(threats)>3: print(f"    ... +{len(threats)-3} more")
                rp=os.path.join(Config.CAMPAIGN_DATA,f"campaign_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt")
                with open(rp,'w',encoding='utf-8') as f: f.write(campaign_det.generate_campaign_report(campaigns))
                Colors.print_success(f"✓ Report saved: {os.path.basename(rp)}")
                valid=[d for d in collected_dnas if d.combined_embedding is not None]
                if len(valid)>=3:
                    try:
                        Colors.print_info("Generating t-SNE visualization...")
                        X2=TSNE(n_components=2,random_state=Config.RANDOM_STATE,
                                perplexity=min(30,len(valid)-1)).fit_transform(np.array([d.combined_embedding for d in valid]))
                        plt.figure(figsize=(12,8))
                        for cid,threats in campaigns.items():
                            idx=[valid.index(t) for t in threats if t in valid]
                            if not idx: continue
                            pts=X2[idx]
                            col='gray' if cid=='UNCLUSTERED' else plt.cm.Set1(hash(cid)%9)
                            plt.scatter(pts[:,0],pts[:,1],c=[col],label=cid,alpha=0.7 if cid!='UNCLUSTERED' else 0.3,s=80,edgecolors='black' if cid!='UNCLUSTERED' else 'none')
                        plt.title('Campaign Clustering (t-SNE)',fontsize=14,fontweight='bold')
                        plt.xlabel('Dim 1'); plt.ylabel('Dim 2'); plt.legend(bbox_to_anchor=(1.05,1))
                        plt.grid(alpha=0.3); plt.tight_layout()
                        vf=os.path.join(Config.CAMPAIGN_DATA,f"campaign_viz_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png")
                        plt.savefig(vf,dpi=Config.VIZ_DPI,bbox_inches='tight'); plt.close()
                        Colors.print_success(f"✓ Visualization: {os.path.basename(vf)}")
                        try: webbrowser.open(f'file://{os.path.abspath(vf)}')
                        except: pass
                    except Exception as e:
                        Colors.print_warning(f"Visualization failed: {e}")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            # ── 13: DNA Database Stats ──
            elif choice=='13':
                Colors.print_header("📊 DNA DATABASE STATS")
                print(f"\n  DNA Samples     : {len(collected_dnas)}")
                print(f"  Vector DB       : {vector_db.get_threat_count() if vector_db else 'N/A'} entries")
                if collected_dnas:
                    we=sum(1 for d in collected_dnas if d.combined_embedding is not None)
                    ws=sum(1 for d in collected_dnas if d.screenshot_path)
                    wd=sum(1 for d in collected_dnas if d.dom_snapshot)
                    n=len(collected_dnas)
                    print(f"  With Embeddings : {we} ({we/n*100:.1f}%)")
                    print(f"  With Screenshots: {ws} ({ws/n*100:.1f}%)")
                    print(f"  With DOM        : {wd} ({wd/n*100:.1f}%)")
                    print(f"  Avg Risk Score  : {np.mean([d.risk_score for d in collected_dnas]):.1f}/100")
                    hi=sum(1 for d in collected_dnas if d.risk_score>=Config.HIGH_RISK)
                    me=sum(1 for d in collected_dnas if Config.MEDIUM_RISK<=d.risk_score<Config.HIGH_RISK)
                    lo=n-hi-me
                    print(f"\n  {Colors.FAIL}High:   {hi:3d} ({hi/n*100:5.1f}%){Colors.ENDC}")
                    print(f"  {Colors.WARNING}Medium: {me:3d} ({me/n*100:5.1f}%){Colors.ENDC}")
                    print(f"  {Colors.OKGREEN}Low:    {lo:3d} ({lo/n*100:5.1f}%){Colors.ENDC}")
                    print(f"\n{Colors.BOLD}Recent:{Colors.ENDC}")
                    for d in collected_dnas[-5:]: print(f"  • {d.threat_id}: {d.url[:50]}... ({d.risk_score}/100)")
                else:
                    Colors.print_warning("No DNA samples yet. Use [10] to start.")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

            else:
                Colors.print_warning("Invalid option. Choose 0-13.")
                input(f"\n{Colors.BOLD}Press Enter to continue...{Colors.ENDC}")

    except KeyboardInterrupt:
        print(f"\n\n{Colors.WARNING}Cancelled by user{Colors.ENDC}")
        monitor.stop_monitoring()
    except Exception as e:
        Colors.print_error(f"Fatal error: {e}")
        logger.error(f"Fatal error: {e}", exc_info=True)
        import traceback; traceback.print_exc()
    finally:
        if monitor.is_monitoring:
            monitor.stop_monitoring()


if __name__=="__main__":
    try:
        main()
    except Exception as e:
        print(f"\n{Colors.FAIL}Critical error: {e}{Colors.ENDC}")
        import traceback; traceback.print_exc()
        input("\nPress Enter to exit...")
```
